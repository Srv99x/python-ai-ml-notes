## Statistics Basics for ML
This notebook covers the basics of statistics for machine learning, including descriptive statistics, distributions, correlation and covariance, z-scores and standardization, hypothesis testing, and probability calculations.

### Descriptive Statistics
Descriptive statistics are used to summarize and describe the main features of a dataset. This includes measures of central tendency (mean, median, mode) and measures of spread (variance, standard deviation, interquartile range).

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(42)

data = np.array([35, 42, 48, 50, 52, 55, 58, 60, 65, 100])  # Include outlier

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]

print("=== Descriptive Statistics ===")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode}")
```

The output of this code will give us the mean, median, and mode of the dataset. The mean is sensitive to outliers, while the median is more robust. The mode is the most frequently occurring value in the dataset.
Common mistake: Using the mean when the data is skewed or has outliers. Quick tip: Use the median when the data is skewed or has outliers.

### Spread / Dispersion
Measures of spread or dispersion are used to describe the amount of variation in a dataset. This includes variance, standard deviation, and interquartile range.

```python
variance = np.var(data)
std_dev = np.std(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)

print(f"\nVariance: {variance:.2f}")
print(f"Std Dev: {std_dev:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")
```

The output of this code will give us the variance, standard deviation, and interquartile range of the dataset. The variance and standard deviation are measures of the spread of the data, while the interquartile range is a measure of the spread of the middle 50% of the data.
Common mistake: Not using the interquartile range when the data is skewed or has outliers. Quick tip: Use the interquartile range when the data is skewed or has outliers.

### Percentiles and Quantiles
Percentiles and quantiles are used to describe the distribution of a dataset. This includes the 25th percentile (Q1), 50th percentile (median), and 75th percentile (Q3).

```python
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

print(f"\nQ1 (25th percentile): {q1:.2f}")
print(f"Q3 (75th percentile): {q3:.2f}")
```

The output of this code will give us the 25th percentile (Q1) and 75th percentile (Q3) of the dataset. These values can be used to describe the distribution of the data.
Common mistake: Not using percentiles and quantiles to describe the distribution of the data. Quick tip: Use percentiles and quantiles to describe the distribution of the data.

### Skewness and Kurtosis
Skewness and kurtosis are used to describe the shape of a distribution. Skewness measures the asymmetry of the distribution, while kurtosis measures the tailedness of the distribution.

```python
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)

print(f"\nSkewness: {skewness:.2f} (right-skewed; outlier on right)")
print(f"Kurtosis: {kurtosis:.2f} (heavier tails than normal)")
```

The output of this code will give us the skewness and kurtosis of the dataset. These values can be used to describe the shape of the distribution.
Common mistake: Not using skewness and kurtosis to describe the shape of the distribution. Quick tip: Use skewness and kurtosis to describe the shape of the distribution.

### Distributions
Distributions are used to model the probability of different outcomes in a dataset. This includes the normal distribution, uniform distribution, and exponential distribution.

```python
from scipy.stats import norm

mu = 100  # Mean
sigma = 15  # Standard deviation
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)
pdf = norm.pdf(x, mu, sigma)

print("\n" + "="*60)
print("\n=== Normal Distribution ===")
print(f"Mean: {mu}, Std Dev: {sigma}")
```

The output of this code will give us the probability density function of the normal distribution with the specified mean and standard deviation.
Common mistake: Not using the normal distribution to model continuous data. Quick tip: Use the normal distribution to model continuous data.

### Correlation and Covariance
Correlation and covariance are used to describe the relationship between two variables. This includes the Pearson correlation coefficient and the Spearman correlation coefficient.

```python
x = np.array([1, 2, 3, 4, 5])
y = 2 * x + np.random.randn(5) * 0.5  # y ≈ 2x with noise

cov_matrix = np.cov(x, y)
print("\n" + "="*60)
print("\n=== Covariance and Correlation ===")
print(f"Covariance matrix:\n{cov_matrix}")
print(f"Cov(X, Y): {cov_matrix[0, 1]:.4f}")
```

The output of this code will give us the covariance matrix and the covariance between the two variables. The covariance matrix can be used to describe the relationship between the variables.
Common mistake: Not using the covariance matrix to describe the relationship between variables. Quick tip: Use the covariance matrix to describe the relationship between variables.

### Z-Scores and Standardization
Z-scores and standardization are used to scale variables to comparable units. This includes the z-score formula and the standardization formula.

```python
data = np.array([32, 45, 52, 50, 49, 55, 70, 48, 105, 90])
mean = np.mean(data)
std = np.std(data)

z_scores = (data - mean) / std
print("\n" + "="*60)
print("\n=== Z-Scores ===")
print(f"Original data: {data}")
print(f"Z-scores: {z_scores.round(2)}")
print(f"Extreme outliers (|z| > 3): {data[np.abs(z_scores) > 3]}")
```

The output of this code will give us the z-scores and the extreme outliers in the dataset. The z-scores can be used to identify outliers and to scale variables to comparable units.
Common mistake: Not using z-scores to identify outliers. Quick tip: Use z-scores to identify outliers.

### Hypothesis Testing
Hypothesis testing is used to determine if observed differences are statistically significant or random. This includes the independent t-test and the paired t-test.

```python
group1 = np.array([85, 88, 90, 92, 91, 87])
group2 = np.array([78, 80, 82, 84, 81, 79])

t_stat1, p_value1 = stats.ttest_ind(group1, group2)
print("\n" + "="*60)
print("\n=== Hypothesis Testing ===")
print("Group 1 mean:", group1.mean())
print("Group 2 mean:", group2.mean())
print(f"T-statistic: {t_stat1:.4f}")
print(f"P-value: {p_value1:.4f}")
if p_value1 < 0.05:
    print("Result: Significant difference (reject null hypothesis)")
else:
    print("Result: No significant difference (fail to reject null hypothesis)")
```

The output of this code will give us the t-statistic and p-value of the independent t-test. The p-value can be used to determine if the observed difference is statistically significant.
Common mistake: Not using the p-value to determine statistical significance. Quick tip: Use the p-value to determine statistical significance.

### Probability Calculations
Probability calculations are used to compute likelihoods for decision-making and model interpretation. This includes the birthday problem and Bayes' theorem.

```python
n_people = 23
prob_all_different = 1
for i in range(1, n_people):
    prob_all_different *= (365 - i) / 365
prob_shared = 1 - prob_all_different

print("\n" + "="*60)
print("\n=== Probability ===")
print(f"With {n_people} people, P(≥2 share birthday): {prob_shared:.4f} (~{prob_shared*100:.1f}%)")
```

The output of this code will give us the probability that at least two people share a birthday in a group of 23 people. This probability can be used to make decisions and interpret models.
Common mistake: Not using probability calculations to make decisions. Quick tip: Use probability calculations to make decisions.

### Practical ML Scenarios
Practical ML scenarios include outlier detection, feature scaling, and checking feature independence.

```python
data = np.random.normal(100, 15, 100)
data = np.append(data, [200, -100])  # Add outliers
z_scores = np.abs((data - np.mean(data)) / np.std(data))
outliers = data[z_scores > 3]
print(f"Detected {len(outliers)} outliers: {outliers}")
```

The output of this code will give us the number of outliers detected in the dataset. The outliers can be used to identify extreme values in the data.
Common mistake: Not using z-scores to detect outliers. Quick tip: Use z-scores to detect outliers.

## Key Takeaways

* Normal distribution: most common; test for normality before analysis.
* Correlation: measure relationships; use Pearson or Spearman appropriately.
* Z-scores: standardize and detect outliers.
* Hypothesis tests: determine statistical significance.
* Bayes: update beliefs with evidence (foundation of probabilistic ML).
