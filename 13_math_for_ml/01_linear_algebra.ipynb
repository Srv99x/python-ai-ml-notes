## Linear Algebra with NumPyThis notebook covers the basics of linear algebra using NumPy, including vectors, matrices, and various operations.Understanding these concepts is crucial for working with machine learning algorithms.

### Vectors and Basic OperationsVectors are 1D arrays representing data points or feature values.They can be created using `np.array()` and support basic operations like addition, subtraction, and scalar multiplication.

```python
import numpy as npv1 = np.array([1, 2, 3])v2 = np.array([4, 5, 6])v_sum = v1 + v2  # [5, 7, 9]v_diff = v1 - v2  # [-3, -3, -3]v_scaled = v1 * 2  # [2, 4, 6]print("Vector addition:", v_sum)print("Vector subtraction:", v_diff)print("Scaled vector:", v_scaled)
```

The output of the vector operations shows the result of adding, subtracting, and scaling vectors.A common mistake is trying to perform operations on vectors of different lengths.A quick tip is to use the `np.array()` function to create vectors and the `+`, `-`, and `*` operators for basic operations.

### Dot Product (Inner Product)The dot product measures the similarity between two vectors.It can be computed using the `np.dot()` function or the `@` operator.

```python
u = np.array([1, 2, 3])v = np.array([0, 1, 2])dot_product = np.dot(u, v)dot_product_alt = u @ vprint("Dot product:", dot_product)print("Dot product (@ operator):", dot_product_alt)
```

The dot product is used in machine learning to compute similarity scores between embeddings, weighted sums, and attention mechanisms.A common mistake is trying to compute the dot product of vectors with different lengths.A quick tip is to use the `np.dot()` function or the `@` operator to compute the dot product.

### Vector NormsVector norms measure the magnitude of a vector.The L2 norm (Euclidean distance) and L1 norm (Manhattan distance) are commonly used.

```python
v = np.array([3, 4])l2_norm = np.linalg.norm(v)l1_norm = np.linalg.norm(v, ord=1)v_normalized = v / np.linalg.norm(v)print("L2 norm:", l2_norm)print("L1 norm:", l1_norm)print("Normalized vector:", v_normalized)
```

The output of the vector norms shows the L2 and L1 norms of the vector.A common mistake is trying to normalize a vector with zero magnitude.A quick tip is to use the `np.linalg.norm()` function to compute the L2 and L1 norms.

### Matrices and Matrix OperationsMatrices represent datasets or transformations.Matrix operations include addition, subtraction, and multiplication.

```python
A = np.array([[1, 2], [3, 4]])B = np.array([[5, 6], [7, 8]])C_sum = A + BC_mult = A * Bprint("Matrix addition:\n", C_sum)print("Element-wise multiplication:\n", C_mult)
```

The output of the matrix operations shows the result of adding and multiplying matrices.A common mistake is trying to perform operations on matrices with different shapes.A quick tip is to use the `+` and `*` operators for element-wise operations.

### Matrix MultiplicationMatrix multiplication is a core operation in neural networks and linear transformations.It can be computed using the `np.dot()` function or the `@` operator.

```python
A = np.array([[1, 2, 3], [4, 5, 6]])B = np.array([[7, 8], [9, 10], [11, 12]])C = np.dot(A, B)C_alt = A @ Bprint("Matrix multiplication (2x3 * 3x2 = 2x2):\n", C)
```

The output of the matrix multiplication shows the result of multiplying two matrices.A common mistake is trying to multiply matrices with incompatible shapes.A quick tip is to use the `np.dot()` function or the `@` operator to compute the matrix product.

### TransposeThe transpose of a matrix flips its rows and columns.It can be computed using the `.T` attribute.

```python
A = np.array([[1, 2, 3], [4, 5, 6]])A_T = A.Tprint("Original matrix shape:", A.shape)print("Transposed matrix shape:", A_T.shape)print("Transposed:\n", A_T)
```

The output of the transpose shows the original and transposed matrices.A common mistake is trying to transpose a non-matrix object.A quick tip is to use the `.T` attribute to compute the transpose.

### Determinant and Matrix InverseThe determinant of a matrix checks if it is invertible.The inverse of a matrix solves linear systems.

```python
A = np.array([[1, 2], [3, 4]])det_A = np.linalg.det(A)A_inv = np.linalg.inv(A)print("Determinant:", det_A)print("Inverse:\n", A_inv)
```

The output of the determinant and inverse shows the result of computing these values.A common mistake is trying to compute the inverse of a non-invertible matrix.A quick tip is to use the `np.linalg.det()` and `np.linalg.inv()` functions to compute the determinant and inverse.

### Eigenvalues and EigenvectorsEigenvalues and eigenvectors are used in PCA for dimensionality reduction and in neural network analysis.They can be computed using the `np.linalg.eig()` function.

```python
A = np.array([[4, -2], [-2, 4]])eigenvalues, eigenvectors = np.linalg.eig(A)print("Eigenvalues:", eigenvalues)print("Eigenvectors:\n", eigenvectors)
```

The output of the eigenvalues and eigenvectors shows the result of computing these values.A common mistake is trying to compute the eigenvalues and eigenvectors of a non-square matrix.A quick tip is to use the `np.linalg.eig()` function to compute the eigenvalues and eigenvectors.

### Rank and Matrix DecompositionThe rank of a matrix indicates the number of linearly independent rows or columns.Matrix decomposition reveals the structure of a matrix.

```python
A = np.array([[1, 2, 3], [2, 4, 6], [1, 1, 1]])rank = np.linalg.matrix_rank(A)U, sigma, V_T = np.linalg.svd(A)print("Matrix rank:", rank)print("SVD - Singular values:", sigma)
```

The output of the rank and matrix decomposition shows the result of computing these values.A common mistake is trying to compute the rank or decomposition of a non-matrix object.A quick tip is to use the `np.linalg.matrix_rank()` and `np.linalg.svd()` functions to compute the rank and decomposition.

### Practical ML ScenariosLinear regression and PCA are two examples of machine learning scenarios that use linear algebra.They can be implemented using the concepts learned in this notebook.

```python
X = np.array([[1, 2], [3, 4], [5, 6]])y = np.array([5, 11, 17])w = np.linalg.inv(X.T @ X) @ X.T @ yy_pred = X @ wprint("Linear regression weights:", w)print("Predictions:", y_pred)print("Actual:", y)
```

The output of the linear regression shows the result of computing the weights and predictions.A common mistake is trying to compute the inverse of a non-invertible matrix.A quick tip is to use the `np.linalg.inv()` function to compute the inverse.
