### Random Variables and DistributionsRandom variables represent uncertainty and randomness in data. They can be discrete or continuous.Discrete distributions model binary outcomes, such as coin flips, while continuous distributions model measurements like heights.

```python
import numpy as npfrom scipy import statsimport matplotlib.pyplot as pltnp.random.seed(42)n_flips = 1000coin_flips = np.random.binomial(n=1, p=0.5, size=n_flips)heads_pct = coin_flips.sum() / n_flipsprint("=== Discrete Distribution: Coin Flips ===")print(f"Heads percentage from {n_flips} flips: {heads_pct:.2%}")heights = np.random.normal(loc=170, scale=10, size=1000)print(f"\n=== Continuous Distribution: Heights ===")print(f"Mean: {heights.mean():.2f} cm")print(f"Std: {heights.std():.2f} cm")print(f"Min: {heights.min():.2f} cm, Max: {heights.max():.2f} cm")
```

The output shows the percentage of heads in a series of coin flips and the mean, standard deviation, minimum, and maximum of a set of heights.A common mistake is to confuse discrete and continuous distributions.A quick tip is to use `np.random.binomial` for discrete distributions and `np.random.normal` for continuous distributions.

### Probability Mass Function (PMF) and Probability Density Function (PDF)The PMF gives the probability of each outcome in a discrete distribution, while the PDF gives the probability density at each point in a continuous distribution.

```python
die_rolls = np.random.randint(1, 7, size=1000)for face in range(1, 7):    prob = (die_rolls == face).sum() / len(die_rolls)    print(f"P(die = {face}): {prob:.3f}")from scipy.stats import normx = np.linspace(-3, 3, 100)pdf = norm.pdf(x)cdf = norm.cdf(x)print(f"\n=== Normal Distribution PDF/CDF ===")print(f"PDF at x=0: {norm.pdf(0):.4f}")print(f"CDF at x=0: {norm.cdf(0):.4f}")print(f"CDF at x=1: {norm.cdf(1):.4f}")print(f"CDF at x=2: {norm.cdf(2):.4f}")x_at_95_pct = norm.ppf(0.95)print(f"Value at 95th percentile: {x_at_95_pct:.4f}")
```

The output shows the probability of each outcome in a discrete distribution and the probability density at each point in a continuous distribution.A common mistake is to confuse the PMF and PDF.A quick tip is to use `np.random.randint` for discrete distributions and `scipy.stats.norm.pdf` for continuous distributions.

### Expectations and VarianceThe expectation is the weighted average of a random variable, while the variance measures the spread of the variable.

```python
x_values = np.array([1, 2, 3, 4, 5, 6])probabilities = np.array([1/6] * 6)expected_value = (x_values * probabilities).sum()print(f"\n=== Expectation and Variance ===")print(f"E[fair die]: {expected_value:.2f}")x_continuous = np.linspace(-10, 10, 1000)pdf_values = norm.pdf(x_continuous, loc=5, scale=2)expected = (x_continuous * pdf_values).mean()print(f"E[Normal(μ=5, σ=2)]: {expected:.2f}")samples = np.random.normal(loc=100, scale=15, size=1000)variance = np.var(samples)std_dev = np.std(samples)print(f"Var[X]: {variance:.2f}")print(f"Std[X]: {std_dev:.2f}")
```

The output shows the expectation and variance of a random variable.A common mistake is to confuse the expectation and variance.A quick tip is to use `np.mean` for expectation and `np.var` for variance.

### Joint and Conditional ProbabilityJoint probability gives the probability of two events happening together, while conditional probability gives the probability of one event given another.

```python
P_rain = 0.3P_go_given_rain = 0.2P_rain_and_go = P_rain * P_go_given_rainprint(f"\n=== Joint and Conditional Probability ===")print(f"P(rain) = {P_rain}")print(f"P(go|rain) = {P_go_given_rain}")print(f"P(rain AND go) = {P_rain_and_go}")P_spam = 0.1P_word_given_spam = 0.8P_word_given_ham = 0.05P_word = (P_word_given_spam * P_spam + P_word_given_ham * (1 - P_spam))P_spam_given_word = (P_word_given_spam * P_spam) / P_wordprint(f"\nSpam Detection:")print(f"P(spam) = {P_spam}")print(f"P('Buy' in email | spam) = {P_word_given_spam}")print(f"P('Buy' in email | ham) = {P_word_given_ham}")print(f"P(spam | 'Buy' in email) = {P_spam_given_word:.4f}")
```

The output shows the joint and conditional probability of two events.A common mistake is to confuse joint and conditional probability.A quick tip is to use the formula `P(A and B) = P(A|B) * P(B)` for joint probability and `P(A|B) = P(A and B) / P(B)` for conditional probability.

### Sampling and Monte CarloMonte Carlo methods simulate complex systems by generating random samples from a distribution.

```python
n_samples = 100000x = np.random.uniform(-1, 1, n_samples)y = np.random.uniform(-1, 1, n_samples)distances = np.sqrt(x**2 + y**2)in_circle = distances <= 1pi_estimate = 4 * in_circle.sum() / n_samplesprint(f"\n=== Monte Carlo Estimation ===")print(f"Estimated π: {pi_estimate:.4f}")print(f"Error: {abs(pi_estimate - np.pi):.4f}")normal_samples = np.random.normal(loc=0, scale=1, size=10000)extreme_events = (normal_samples > 5).sum()print(f"\nP(X > 5) with normal sampling: {extreme_events / 10000}")shifted_samples = np.random.normal(loc=5, scale=1, size=10000)weights = norm.pdf(shifted_samples) / norm.pdf(shifted_samples, loc=5)p_extreme_importance = (weights[shifted_samples > 5].sum() / weights.sum())print(f"P(X > 5) with importance sampling: {p_extreme_importance:.6f}")
```

The output shows the estimated value of π using Monte Carlo methods.A common mistake is to confuse the concept of sampling and Monte Carlo methods.A quick tip is to use `np.random.uniform` for uniform sampling and `scipy.stats.norm.pdf` for normal sampling.

### Probability Distributions for MLDifferent distributions model different phenomena in machine learning.

```python
p_heads = 0.6bernoulli_trials = np.random.binomial(n=1, p=p_heads, size=1000)print(f"\n=== Common Distributions for ML ===")print(f"Bernoulli(p={p_heads}): {bernoulli_trials.mean():.3f}")successes = np.random.binomial(n=10, p=0.5, size=1000)print(f"Binomial(n=10, p=0.5) mean: {successes.mean():.2f}")events = np.random.poisson(lam=3, size=1000)print(f"Poisson(λ=3) mean: {events.mean():.2f}")waiting_times = np.random.exponential(scale=2, size=1000)print(f"Exponential(λ=0.5) mean: {waiting_times.mean():.2f}")alpha = [1, 1, 1]probabilities_dist = np.random.dirichlet(alpha, size=5)print(f"Dirichlet (normalized): each row sums to 1")print(f"Sample:\n{probabilities_dist.round(3)}")
```

The output shows the mean of different distributions.A common mistake is to confuse the different distributions.A quick tip is to use `np.random.binomial` for Bernoulli and binomial distributions, `np.random.poisson` for Poisson distribution, `np.random.exponential` for exponential distribution, and `np.random.dirichlet` for Dirichlet distribution.

### Practical ML ScenariosThese scenarios demonstrate how to apply probability concepts in machine learning.

```python
predicted_scores = np.array([0.1, 0.8, 0.9, 0.2, 0.7, 0.4])probabilities = 1 / (1 + np.exp(-predicted_scores))predictions = (probabilities > 0.5).astype(int)print("Classification predictions:")for i, (score, prob, pred) in enumerate(zip(predicted_scores, probabilities, predictions)):    print(f"  Sample {i}: score={score:.2f}, P(class=1)={prob:.3f}, prediction={pred}")sample_mean = 100sample_std = 15confidence = 0.95z_critical = norm.ppf((1 + confidence) / 2)margin_of_error = z_critical * (sample_std / np.sqrt(100))ci_lower, ci_upper = sample_mean - margin_of_error, sample_mean + margin_of_errorprint(f"\n95% Confidence interval for mean: [{ci_lower:.2f}, {ci_upper:.2f}]")observations = np.array([10, 12, 11, 13, 12, 11, 14])likelihood_poisson_3 = np.prod(stats.poisson.pmf(observations, mu=3))likelihood_poisson_12 = np.prod(stats.poisson.pmf(observations, mu=12))print(f"\nLikelihood (μ=3): {likelihood_poisson_3:.2e}")print(f"Likelihood (μ=12): {likelihood_poisson_12:.2e}")
```

The output shows the classification predictions, confidence interval, and likelihood of different models.A common mistake is to confuse the different concepts.A quick tip is to use `scipy.stats.norm.pdf` for normal distribution, `scipy.stats.poisson.pmf` for Poisson distribution, and `np.prod` for calculating likelihood.
