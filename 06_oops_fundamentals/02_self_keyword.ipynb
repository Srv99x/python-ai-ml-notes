## Self Keyword and Instance Context in Methods
This notebook covers the self keyword and instance context in methods. It explains how to use the self parameter to access instance attributes and methods.
We will create a Student class and demonstrate how to use the self keyword to access and modify instance attributes.

### Class Definition
A class is defined using the `class` keyword followed by the name of the class. In this case, we define a `Student` class.
The class has class attributes and methods.

```python
class Student:
    # Class attributes
    name = "Sourav"
    language = "Js"
    rollNo = 14
    sec = 'A'

    def __init__(self):         # Dunder method, it is automatically called when an object is created
        print("I'm creating an object!")

    def getInfo(self):
        print(f"The name  is {self.name}.\nThe language is {self.language}")

    @staticmethod
    def greet():
        print("Good morning")
```

### Creating an Object and Accessing Attributes
We create an object `info` of the `Student` class and access its attributes. We also demonstrate how to modify an instance attribute.
Note that when we access an attribute, Python first looks for an instance attribute with that name, and if not found, it looks for a class attribute.

```python
info = Student()
info.language = "Py"      # object attribute or instance attribute
```

### Output and Explanation
When we run the above code, we see that an object is created and the `__init__` method is called automatically.
One common mistake is to forget that instance attributes are accessed using the `self` parameter.
A quick tip is to use the `print` function to verify the values of instance attributes.

### Using the Self Parameter
The `self` parameter is used to access instance attributes and methods. We demonstrate how to use the `self` parameter in the `getInfo` method.
Note that the `self` parameter is passed automatically when we call an instance method.

```python
info.getInfo()    # self parameter
# Student.getInfo(info)     # self parameter
```

### Output and Explanation
When we run the above code, we see that the `getInfo` method is called and it prints the values of the instance attributes.
One common mistake is to forget to pass the `self` parameter when calling an instance method using the class name.
A quick tip is to use the `@staticmethod` decorator to define a static method that does not require the `self` parameter.

### Static Methods
A static method is a method that belongs to a class rather than an instance. We demonstrate how to define and call a static method.
Note that a static method does not have access to instance attributes.

```python
info.greet()     
```

### Output and Explanation
When we run the above code, we see that the `greet` method is called and it prints a message.
One common mistake is to try to access instance attributes in a static method.
A quick tip is to use static methods for utility functions that do not depend on instance attributes.

## Key Takeaways
* The `self` parameter is used to access instance attributes and methods.
* Instance attributes are accessed using the `self` parameter.
* Static methods do not have access to instance attributes.
* The `@staticmethod` decorator is used to define a static method.
