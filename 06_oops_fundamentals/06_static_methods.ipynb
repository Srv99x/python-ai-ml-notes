## Static Methods and @staticmethod UsageThis topic covers the concept of static methods in Python, their usage, and the role of the @staticmethod decorator.It explains how static methods can be used to extend the behavior of a function without permanently modifying it.

### What are Static Methods?Static methods are methods that belong to a class rather than an instance of the class.They are used to group related functionality together and can be called without creating an instance of the class.The @staticmethod decorator is used to define a static method.

```python
class Student:    @staticmethod  # decorator: it allows us to wrap another function in order to extend the behavior of the function, without permanently modifying it    def hello():        print("Hello!")
```

### Output and ExplanationWhen we call the hello() method, it will print 'Hello!' to the console.Common mistake: Forgetting to use the @staticmethod decorator when defining a static method.Quick tip: Use static methods when you want to group related functionality together, but don't need to access instance variables.

### Instance MethodsInstance methods are methods that belong to an instance of a class.They are used to perform operations on instance variables and can be called only after creating an instance of the class.

```python
    def __init__(self, name, marks):        self.name = name        self.marks = marks    def get_sum(self):        sum = 0        for val in self.marks:            sum += val        print(f"Hello {self.name} your average marks is {sum/len(self.marks)}")
```

### Output and ExplanationWhen we create an instance of the Student class and call the get_sum() method, it will print a message with the student's name and average marks.Common mistake: Forgetting to use the self parameter when defining an instance method.Quick tip: Use instance methods when you need to perform operations on instance variables.

```python
s1 = Student("Sourav", [78, 89, 97])s1.get_sum()
```

## Key Takeaways* Static methods belong to a class and can be called without creating an instance of the class.* The @staticmethod decorator is used to define a static method.* Instance methods belong to an instance of a class and can be called only after creating an instance of the class.* Use static methods when you want to group related functionality together, but don't need to access instance variables.* Use instance methods when you need to perform operations on instance variables.
