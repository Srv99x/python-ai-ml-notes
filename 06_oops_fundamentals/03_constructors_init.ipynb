## Constructors and Object Initialization with __init__
This topic covers the basics of constructors in Python, specifically the __init__ method, and how it is used for object initialization.
We will explore how to define and use constructors to create objects with specific attributes.

### What is a Constructor?
A constructor is a special method in a class that is automatically called when an object is created from the class.
In Python, the constructor method is defined using the __init__ method.

```python
class Employee:
    # Default constructor
    def __init_subclass__(self):
        pass

    # Parameterized constructor
    def __init__(self, name, salary, language):  # Dunder method, it is automatically called when an object is created
        self.name = name  # The self parameter is a reference to the current instance of the class, and is used to access variables that belong to the class
        self.salary = salary
        self.language = language
        print("I'm creating an object")
```

### Output and Explanation
When an object is created from the Employee class, the __init__ method is automatically called, printing "I'm creating an object".
Common mistake: Forgetting to use the self parameter to access class variables.
Quick tip: Use the __init__ method to initialize objects with specific attributes.

### Defining Methods
Methods are functions that belong to a class and can be used to perform actions on objects.
In this example, we define two methods: welcome and get_salary.

```python
    def welcome(self):  # Methods
        print("Hello Employee!", self.name)

    def get_salary(self):
        return self.salary
```

### Using Methods
We can use the methods defined in the Employee class to perform actions on objects.
Common mistake: Forgetting to use the self parameter when defining methods.
Quick tip: Use methods to encapsulate actions that can be performed on objects.

```python
e1 = Employee("Harry", 130000, "Java")
# Harry.name = "Harry"
print(f"Name: {e1.name}\nLanguage: {e1.language}\nSalary: {e1.salary}")
e1.welcome()
print(f"The salary is: {e1.get_salary()}")
```

### Creating Multiple Objects
We can create multiple objects from the Employee class, each with its own attributes.
Common mistake: Forgetting to pass the required arguments when creating an object.
Quick tip: Use the __init__ method to initialize objects with specific attributes.

```python
e2 = Employee("Rohan", 135000, "C++")
print(f"Name: {e2.name}\nLanguage: {e2.language}\nSalary {e2.salary}")
e2.welcome()
print(f"The salary is: {e2.get_salary()}")
```

## Key Takeaways
* Constructors are special methods that are automatically called when an object is created.
* The __init__ method is used to define a constructor in Python.
* Methods are functions that belong to a class and can be used to perform actions on objects.
* Use the self parameter to access class variables and define methods.
