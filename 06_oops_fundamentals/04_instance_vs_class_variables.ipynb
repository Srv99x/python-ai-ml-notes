## Instance Variables vs Class Variables
This topic covers the difference between instance variables and class variables in Python.
We will explore how these variables interact and affect the behavior of objects and classes.

### Instance Variables vs Class Variables
In Python, instance variables have more precedence than class variables.
This means that if an instance variable and a class variable have the same name, the instance variable will be used.

```python
# Define a class Employee with class variables language and name
class Employee:
    language = "Python"  # class variable
    name = "sourav"  # class variable

# Create an instance of the Employee class
info = Employee()

# Assign a new value to the language attribute of the instance
info.language = "JavaScript"  # instance variable

# Print the language and name attributes of the instance
print(info.language, info.name)
```

### Output and Explanation
The output of the code will be: JavaScript sourav
This is because the instance variable `language` has more precedence than the class variable `language`.
A common mistake is to assume that the class variable will be used when an instance variable with the same name is not defined.
A quick tip is to use the `self` parameter to access instance variables and the `classname` to access class variables.

## Key Takeaways
* Instance variables have more precedence than class variables.
* Use the `self` parameter to access instance variables.
* Use the `classname` to access class variables.
