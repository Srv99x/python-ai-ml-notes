## Properties with @property getter/setter patterns
This topic covers the use of properties in Python classes, specifically focusing on the @property decorator for getter methods. It demonstrates how to calculate and display a student's percentage score using a property.

### Defining a Class with Properties
We define a class `Student` with an initializer method `__init__` that takes scores for physics, chemistry, and math. We then use the `@property` decorator to create a getter method `calcPercentage` that calculates the average score.

```python
class Student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math

    @property
    def calcPercentage(self):
        # Calculate the average score and return it as a string with a '%' symbol
        return str((self.phy + self.chem + self.math)/3) + "%"
```

### Understanding the Output and Common Mistakes
When we create an instance of the `Student` class and access the `calcPercentage` property, it calculates and returns the average score. A common mistake is to try to call the property like a method (using parentheses), which will result in an error. A quick tip is to use properties to encapsulate calculations or data transformations that depend on the object's state.

### Creating an Instance and Accessing the Property
We create an instance of the `Student` class with scores for physics, chemistry, and math, and then access the `calcPercentage` property to calculate and display the average score.

```python
stu1 = Student(98, 89, 95)
print(stu1.calcPercentage)
stu1.phy = 70
print(stu1.calcPercentage)
```

### Output Explanation and Quick Tips
The output will display the average score before and after changing the physics score. A quick tip is to use the `@property` decorator to create read-only properties that can be accessed like attributes, making the code more readable and intuitive.

## Key Takeaways
* Properties in Python classes can be used to encapsulate calculations or data transformations.
* The `@property` decorator is used to create getter methods for properties.
* Properties can be accessed like attributes, making the code more readable and intuitive.
* Changing the object's state can affect the property's value.
