## OOP Fundamentals: Classes, Objects, Methods
This notebook covers the basics of Object-Oriented Programming (OOP) in Python, including classes, objects, and methods.
It provides a simple example to demonstrate how classes and objects are used to organize and structure code.

### Classes and Objects
In OOP, a class is a blueprint or template that defines the properties and behavior of an object.
An object is an instance of a class, and has its own set of attributes (data) and methods (functions).

```python
class Student:
    # Class attributes
    name = "Sourav"              # default name
    dept = "BTech CSE"         # department
    rollNo = 14                 # roll number
    sec = 'A'                   # section

info = Student()               # create an object of the Student class
info.language = "Py"         # object attribute or instance attribute
print(f"{info.name}\n{info.dept}\n{info.rollNo}\n{info.sec}\n{info.language}")
```

### Output and Explanation
The output of the above code will be:
Sourav
BTech CSE
14
A
Py
One common mistake is to confuse class attributes with object attributes. Class attributes are shared by all objects of the class, while object attributes are unique to each object.
A quick tip is to use the `self` parameter to access and modify object attributes within methods.

## Key Takeaways
* Classes define the properties and behavior of objects.
* Objects are instances of classes, with their own attributes and methods.
* Class attributes are shared by all objects of the class, while object attributes are unique to each object.
