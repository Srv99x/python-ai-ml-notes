# NumPy: Reshaping, Stacking, and Splitting Arrays

Manipulating the shape and structure of arrays is a fundamental skill in data processing. NumPy provides a rich set of functions to reshape, combine, and split arrays.

## 1. Reshaping Arrays

The `reshape()` method allows you to change the shape (the number of elements in each dimension) of an array without changing its data. The new shape must be compatible with the original shape, meaning it must have the same total number of elements.

```python
import numpy as np

# Create a 1D array with 12 elements
arr = np.arange(12)
print("Original array:\n", arr)

# Reshape it to a 3x4 matrix
reshaped_arr = arr.reshape(3, 4)
print("\nReshaped to 3x4:\n", reshaped_arr)
```

### The -1 Argument

You can use `-1` as a placeholder for one dimension, and NumPy will automatically infer its size based on the other dimensions and the total number of elements.

### Reshape with -1 (Automatic Dimension Inference)
When you know the total elements but not one dimension, use -1 to auto-compute.
Syntax: `arr.reshape(new_shape_with_minus_one)`

```python
# Let NumPy figure out the number of columns
reshaped_auto = arr.reshape(2, -1)
print("Reshaped to 2 rows (auto columns):\n", reshaped_auto)
print("New shape:", reshaped_auto.shape)
```

### Flattening an Array

To convert a multi-dimensional array into a 1D array, you can use `ravel()` or `flatten()`.
- `ravel()` returns a *view* of the original array whenever possible (more memory efficient).
- `flatten()` always returns a *copy*.

### Flatten and Ravel
Convert multidimensional arrays to 1D for certain operations.
Syntax: `arr.flatten()` or `arr.ravel()`

```python
arr_2d = np.arange(6).reshape(2, 3)
print("Original 2D array:\n", arr_2d)

# Flatten using ravel
print("\nFlattened with ravel():\n", arr_2d.ravel())
```

## 2. Stacking Arrays

Stacking is used to join a sequence of arrays along a new axis.

### Reshape for Batch Processing
Neural networks often expect (batch_size, features) format.
Syntax: `arr.reshape(batch_size, features)`

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Stack arrays vertically (row-wise)
vertical_stack = np.vstack((a, b))
print("Vertical Stack (vstack):\n", vertical_stack)
print("Shape:", vertical_stack.shape)
```

### Horizontal Stacking

`hstack` joins arrays along the horizontal axis (column-wise).

### Stacking Arrays
Combine multiple arrays along a new or existing axis.
Syntax: `np.vstack()`, `np.hstack()`, or `np.stack()`

```python
a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])

# Stack arrays horizontally (column-wise)
horizontal_stack = np.hstack((a, b))
print("Horizontal Stack (hstack):\n", horizontal_stack)
print("Shape:", horizontal_stack.shape)
```

### `concatenate`

`concatenate` is a more general function that joins a sequence of arrays along an existing axis. You specify the axis with the `axis` argument.
- `axis=0` is like `vstack` for 2D arrays.
- `axis=1` is like `hstack` for 2D arrays.

### Horizontal Stack (Column-wise Concatenation)
Combine multiple feature sets into one matrix.
Syntax: `np.hstack()`

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])

# Concatenate along axis 0 (rows)
concat_0 = np.concatenate((a, b), axis=0)
print("Concatenate (axis=0):\n", concat_0)

# Concatenate along axis 1 (columns) - this will fail due to shape mismatch
try:
    np.concatenate((a, b), axis=1)
except ValueError as e:
    print(f"\nError with axis=1: {e}")

# To make it work, we need to transpose b
concat_1 = np.concatenate((a, b.T), axis=1)
print("\nConcatenate (axis=1) with transpose:\n", concat_1)
```

## 3. Splitting Arrays

Splitting is the reverse of stacking. You can split an array into multiple smaller arrays.

### Stack along New Axis
Create 3D structures for batch operations on multiple matrices.
Syntax: `np.stack()`

```python
arr = np.arange(9)

# Split into 3 equal-sized arrays
split_arrs = np.split(arr, 3)
print("Original array:", arr)
print("\nSplit into 3 arrays:\n", split_arrs)
```

### Splitting at specific points

You can also pass a list of indices where the array should be split.

### Concatenate along Existing Axis
More flexible than vstack/hstack for arbitrary axis specification.
Syntax: `np.concatenate()`

```python
# Split at index 3 and index 5
split_points = np.split(arr, [3, 5])
print("Split at indices 3 and 5:\n", split_points)
```

### `hsplit` and `vsplit`

Similar to `hstack` and `vstack`, there are helper functions `hsplit` (splits along the horizontal axis) and `vsplit` (splits along the vertical axis).

### Splitting Arrays
Split data into train/test or stratified subsets for model evaluation.
Syntax: `np.array_split()`

```python
arr_2d = np.arange(16).reshape(4, 4)
print("Original 2D array:\n", arr_2d)

# Vertically split into 2 arrays
vsplit_arrs = np.vsplit(arr_2d, 2)
print("\nVertically split:\n", vsplit_arrs)

# Horizontally split into 4 arrays
hsplit_arrs = np.hsplit(arr_2d, 4)
print("\nHorizontally split:\n", hsplit_arrs)
```

## Cheat Sheet: Reshaping, Stacking & Splitting

| Function | Description | Example |
| :--- | :--- | :--- |
| **`arr.reshape(new_shape)`** | Reshapes an array without changing its data. Total elements must remain the same. | `arr.reshape(4, 3)` |
| **`arr.reshape(n, -1)`** | A powerful feature where NumPy infers one of the dimensions. | `arr.reshape(2, -1)` |
| **`arr.ravel()`** | Returns a flattened (1D) **view** of the array. Fast and memory-efficient. | `arr_2d.ravel()` |
| **`arr.flatten()`** | Returns a flattened (1D) **copy** of the array. Safer if you plan to modify it. | `arr_2d.flatten()` |
| **`np.vstack((a, b))`** | Stacks arrays in sequence vertically (row-wise). | `np.vstack((arr1, arr2))` |
| **`np.hstack((a, b))`** | Stacks arrays in sequence horizontally (column-wise). | `np.hstack((arr1, arr2))` |
| **`np.concatenate((a,b), axis)`**| The general-purpose joining function. `axis=0` for vertical, `axis=1` for horizontal. | `np.concatenate((a, b), axis=0)` |
| **`np.split(arr, sections)`** | Splits an array into multiple sub-arrays. Can be an integer for equal parts or a list of indices. | `np.split(arr, 3)` or `np.split(arr, [2, 5])` |
| **`np.vsplit(arr, sections)`** | Splits an array into multiple sub-arrays vertically (row-wise). | `np.vsplit(arr_2d, 2)` |
| **`np.hsplit(arr, sections)`** | Splits an array into multiple sub-arrays horizontally (column-wise). | `np.hsplit(arr_2d, 4)` |
| **`arr.T`** | Transposes an array, swapping its axes. Rows become columns and vice-versa. | `arr_2d.T` |

**View vs. Copy Quick Reference:**
- **View (Shallow Copy):** `reshape`, `ravel`, slicing (`arr[:]`). Changes to the new array **will** affect the original.
- **Copy (Deep Copy):** `flatten`, `.copy()`, functions that create new arrays (`zeros`, `ones`, etc.). Changes to the new array **will not** affect the original.

## 4. Common Mistakes & Takeaways

1.  **Incompatible Reshape**: Trying to reshape an array into a shape that doesn't have the same number of total elements will cause a `ValueError`.
2.  **`ravel()` vs `flatten()`**: Forgetting that `ravel()` can return a view can lead to bugs where modifying the flattened array unintentionally changes the original array. Use `flatten()` if you are sure you need a copy.
3.  **Stacking/Concatenating Shapes**: When stacking, make sure the arrays have compatible shapes along the dimensions you are *not* joining. For `concatenate`, the dimensions must match along the axis you are *not* joining on.
4.  **Splitting Unevenly**: If you try to split an array into an unequal number of parts without providing specific indices, you will get a `ValueError`.

**Takeaway**: These functions are the building blocks for cleaning and preparing data for machine learning models. Being comfortable with reshaping, stacking, and splitting is crucial for effective data manipulation.
