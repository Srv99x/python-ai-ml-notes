# Plot Customization

A good plot is not just correct, it's also easy to understand. Customizing your plots with colors, styles, and legends is key to effective communication.

## 1. Colors, Line Styles, and Markers

You can control the appearance of your plot elements directly in the plotting functions.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)

fig, ax = plt.subplots()

# Customize color, linestyle, and marker
ax.plot(x, x, color='red', linestyle='--', marker='o', label='y = x')
ax.plot(x, x**2, color='#33FF57', linestyle=':', marker='x', label='y = x^2') # Hex color
ax.plot(x, 2*x + 5, color='blue', linestyle='-.', marker='s', label='y = 2x+5')

# Add a legend to explain the lines
ax.legend()

ax.set_title("Customized Plot")
plt.show()
```

### Common Options

-   **`color`**: Can be a name (`'red'`), a hex code (`'#FF5733'`), or an RGB tuple.
-   **`linestyle`**: `'-'` (solid), `'--'` (dashed), `':'` (dotted), `'-.'` (dash-dot).
-   **`marker`**: `'.'` (point), `'o'` (circle), `'x'`, `'s'` (square), `'*'` (star).

## 2. Adding a Legend

A legend is crucial for plots with multiple lines or groups. To create one:
1.  Add a `label` to each plot element you want in the legend (e.g., `ax.plot(..., label='My Line')`).
2.  Call `ax.legend()`.

## 3. Adjusting Figure Size

Sometimes the default plot size is too small. You can set it when you create the figure using the `figsize` argument in `plt.subplots()`.

```python
import matplotlib.pyplot as plt

# figsize=(width, height) in inches
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot([1, 2, 3], [3, 1, 2])
ax.set_title("A Wider Plot")
plt.show()
```

## Cheat Sheet: Plot Customization

| Task | Function / Parameter | Example | Notes |
| :--- | :--- | :--- | :--- |
| **Set Color** | `color` | `ax.plot(x, y, color='green')` | Names, hex codes, etc. |
| **Set Line Style** | `linestyle` | `ax.plot(x, y, linestyle='--')` | `'-'`, `'--'`, `':'`, `'-.'` |
| **Set Marker** | `marker` | `ax.plot(x, y, marker='o')` | `'.'`, `'o'`, `'x'`, `'s'` |
| **Add Legend** | 1. `label` param<br>2. `ax.legend()` | `ax.plot(..., label='A')`<br>`ax.legend()` | `label` is required for `legend` to work. |
| **Set Figure Size** | `figsize` in `subplots` | `fig, ax = plt.subplots(figsize=(10, 5))` | (width, height) in inches. |
| **Set Axis Limits** | `ax.set_xlim()` / `ax.set_ylim()` | `ax.set_xlim(0, 10)` | Controls the visible data range. |
| **Add Grid** | `ax.grid(True)` | `ax.grid(True, linestyle='--', alpha=0.6)` | Helps in reading values. |
