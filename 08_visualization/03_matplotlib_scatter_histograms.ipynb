# Scatter Plots and Histograms

Let's explore two more essential plot types: **scatter plots** for visualizing the relationship between two variables, and **histograms** for understanding the distribution of a single variable.

## 1. Scatter Plots

A scatter plot uses dots to represent the values obtained for two different variables. It's the best way to visualize the relationship or correlation between them.

The function is `ax.scatter(x_values, y_values)`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some correlated data
np.random.seed(42) # for reproducibility
x = np.random.rand(50) * 10
y = 2 * x + 1 + np.random.randn(50) * 2 # y = 2x + 1 + some noise

# Create the plot
fig, ax = plt.subplots()
ax.scatter(x, y)

# Add labels
ax.set_xlabel("Independent Variable (X)")
ax.set_ylabel("Dependent Variable (Y)")
ax.set_title("Scatter Plot of X vs Y")

plt.show()
```

## 2. Histograms

A histogram is used to visualize the distribution of a single numerical variable. It groups numbers into ranges (called "bins") and the height of the bar shows how many data points fall into that range.

The function is `ax.hist(data, bins=number_of_bins)`.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data from a normal distribution (bell curve)
data = np.random.randn(1000)

# Create the plot
fig, ax = plt.subplots()
# 'bins=30' means we want to group the data into 30 ranges
ax.hist(data, bins=30)

# Add labels
ax.set_xlabel("Value")
ax.set_ylabel("Frequency")
ax.set_title("Histogram of a Normal Distribution")

plt.show()
```

### Common Mistakes & Takeaways

1.  **Choosing the Right Number of Bins**: For histograms, the number of bins can significantly change the appearance of the plot. Too few bins can hide important details, while too many can make the plot noisy. It often takes some experimentation to find a good value.
2.  **Over-plotting in Scatter Plots**: If you have a huge amount of data, a scatter plot can become a dense, unreadable blob. In such cases, you can adjust the marker size (`s=...`), transparency (`alpha=...`), or use a different plot type like a 2D histogram.

**Takeaway**: Use `ax.scatter()` to check for relationships between two variables. Use `ax.hist()` to understand the shape and distribution of a single variable.

## Cheat Sheet: Scatter & Histograms

| Plot Type | Function | Use Case | Notes |
| :--- | :--- | :--- | :--- |
| **Scatter Plot** | `ax.scatter(x, y)` | Visualizing the relationship between two continuous variables. | Look for patterns, correlations, and outliers. |
| **Histogram** | `ax.hist(data, bins=N)` | Understanding the distribution of a single numerical variable. | `bins` controls the number of bars. |
| **`alpha`** | `ax.scatter(x, y, alpha=0.5)` | A useful parameter to set transparency (0-1) for overlapping points. | Helps with over-plotting. |
| **`s`** | `ax.scatter(x, y, s=10)` | Sets the marker size in a scatter plot. | Can also be an array to give each point a different size. |
