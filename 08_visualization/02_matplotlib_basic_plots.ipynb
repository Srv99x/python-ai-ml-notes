# Basic Plotting with Matplotlib

Now that we understand the Figure/Axes structure, let's create the two most common types of plots: **line plots** and **bar charts**.

## 1. Line Plots

A line plot is used to visualize the relationship between two continuous variables. It's excellent for showing trends over time or how one variable changes with another.

The function is `ax.plot(x_values, y_values)`.

```python
import matplotlib.pyplot as plt

# Data for a simple line plot
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [100, 120, 150, 140, 180, 200]

# Create the plot
fig, ax = plt.subplots()
ax.plot(years, sales)

# Add labels for clarity
ax.set_xlabel("Year")
ax.set_ylabel("Sales (in thousands)")
ax.set_title("Company Sales Over Time")

plt.show()
```

## 2. Bar Charts

A bar chart is used to compare quantities among different categories. It uses rectangular bars with heights or lengths proportional to the values they represent.

The function is `ax.bar(categories, values)`.

```python
import matplotlib.pyplot as plt

# Data for a bar chart
products = ['A', 'B', 'C', 'D']
units_sold = [250, 180, 320, 90]

# Create the plot
fig, ax = plt.subplots()
ax.bar(products, units_sold)

# Add labels
ax.set_xlabel("Product Category")
ax.set_ylabel("Units Sold")
ax.set_title("Sales by Product Category")

plt.show()
```

### Common Mistakes & Takeaways

1.  **Using the Wrong Plot Type**: Don't use a line plot for categorical data. A line implies a continuous connection between points, which doesn't make sense for distinct categories like 'Product A' and 'Product B'. Use bar charts for categorical comparisons.
2.  **Forgetting Labels**: A plot without labels is meaningless. Always label your x-axis, y-axis, and give your plot a descriptive title using `ax.set_xlabel()`, `ax.set_ylabel()`, and `ax.set_title()`.

**Takeaway**: Choose your plot type based on your data. Use `ax.plot()` for continuous data (like time-series) and `ax.bar()` for comparing categorical data.

## Cheat Sheet: Basic Plots

| Plot Type | Function | Use Case | Notes |
| :--- | :--- | :--- | :--- |
| **Line Plot** | `ax.plot(x, y)` | Showing trends in continuous data over time. | `x` and `y` should be array-like. |
| **Bar Chart** | `ax.bar(categories, values)` | Comparing quantities across different categories. | `categories` are typically strings. |
| **Set X-Label** | `ax.set_xlabel("Label Text")` | Adds a label to the horizontal axis. | Essential for context. |
| **Set Y-Label** | `ax.set_ylabel("Label Text")` | Adds a label to the vertical axis. | Essential for context. |
| **Set Title** | `ax.set_title("Title Text")` | Adds a title to the top of the plot. | Essential for context. |
