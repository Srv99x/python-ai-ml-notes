# Introduction to Matplotlib: Figures and Axes

Welcome to data visualization! Matplotlib is the most widely used library in Python for creating static, animated, and interactive visualizations.

To understand Matplotlib, we must first understand its core components: the **Figure** and the **Axes**.

### The Figure: The Canvas

Think of the `Figure` as the entire window or page that everything is drawn on. It's the top-level container for all the plot elements. A single figure can contain multiple plots.

### The Axes: The Plot

An `Axes` is what we typically think of as 'a plot'. It's the region of the image with the data space. It contains the x-axis, y-axis, labels, and the marks representing the data (lines, bars, points, etc.).

**Analogy**: The `Figure` is the picture frame, and each `Axes` is a photo inside that frame. You can have a frame with just one photo, or a frame with multiple photos arranged in a grid.

### Creating a Plot

The most common way to start is by using `plt.subplots()`. This handy function creates a `Figure` and one or more `Axes` objects all at once.

```python
import matplotlib.pyplot as plt

# Create a Figure and a single Axes (the plot)
fig, ax = plt.subplots()

# The 'ax' object is now our canvas to draw on.
# Let's plot some simple data.
x_values = [1, 2, 3, 4, 5]
y_values = [2, 3, 5, 7, 11]

ax.plot(x_values, y_values)

# Display the figure
plt.show()
```

### Code Explanation

1.  **`import matplotlib.pyplot as plt`**: This is the standard convention for importing Matplotlib's Pyplot interface.
2.  **`fig, ax = plt.subplots()`**: This creates a new figure (`fig`) and a single subplot or axes (`ax`) within that figure.
3.  **`ax.plot(...)`**: This is the core plotting command. We call the `.plot()` method *on our axes object* (`ax`) to draw the data.
4.  **`plt.show()`**: This command displays the figure that we've created. In many environments (like Jupyter notebooks), the plot might show up anyway, but `plt.show()` is the explicit and correct way to render it.

### Common Mistakes & Takeaways

1.  **Confusing `plt` and `ax`**: A common beginner mistake is mixing up the state-based interface (`plt.plot()`) and the object-oriented interface (`ax.plot()`). While `plt.plot()` can work for simple plots, it's less explicit and can be confusing when you have multiple plots. **Best Practice**: Always use the object-oriented approach (`fig, ax = plt.subplots()` and `ax.plot()`) as it's more powerful and clearer.
2.  **Forgetting `plt.show()`**: In regular Python scripts (not notebooks), your plot will not appear unless you explicitly call `plt.show()`.

**Takeaway**: The fundamental structure of a Matplotlib visualization is a `Figure` that contains one or more `Axes`. By creating and manipulating these objects, you have full control over your plots.
