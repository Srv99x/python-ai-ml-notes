### ⚠️ Common Mistake: Forgetting `plt.axis('scaled')` or `ax.set_aspect('equal')`

If you don't set an equal aspect ratio, Matplotlib will scale the x and y axes independently to fit the available space. This will distort your shapes, making circles look like ellipses.

### ✅ Key Takeaway

The process for drawing any shape is generally the same:
1.  Get the current `Axes` (`plt.gca()`).
2.  Create a `patch` object (e.g., `plt.Circle(...)`).
3.  Add the patch to the `Axes` (`ax.add_patch(...)`).
4.  Set the aspect ratio to `'equal'` or `'scaled'`.
5.  Show the plot (`plt.show()`).

### 📋 Complete Cheat Sheet

```python
# === Drawing a Circle with Matplotlib ===

import matplotlib.pyplot as plt

# 1. Define the parameters for the circle
center_coordinates = (0, 0)
radius = 5
face_color = 'lightblue'
edge_color = 'darkblue'

# 2. Get the current Axes object
#    'gca' stands for "Get Current Axes". It gives you the main plotting area.
ax = plt.gca()

# 3. Create a Circle "patch"
#    A patch is a 2D artist object with a face (color) and an edge (color).
#    - The first argument is the (x, y) center of the circle.
#    - `radius` is the circle's radius.
#    - `fc` is the facecolor (the fill).
#    - `ec` is the edgecolor (the border).
circle = plt.Circle(center_coordinates, radius, fc=face_color, ec=edge_color, linewidth=2)

# 4. Add the patch to the Axes
#    This tells Matplotlib to include the circle in the plot.
ax.add_patch(circle)

# 5. Set the aspect ratio to be equal
#    CRITICAL STEP: This ensures the x and y axes are scaled the same way,
#    so the circle looks like a circle and not an oval (ellipse).
ax.set_aspect('equal', adjustable='box')

# 6. Set the plot limits (optional but good practice)
#    This ensures the entire circle is visible in the final plot.
ax.set_xlim(-radius - 1, radius + 1)
ax.set_ylim(-radius - 1, radius + 1)

# 7. Add labels and a title
plt.grid(True, linestyle='--', alpha=0.6)
plt.title("Drawing a Circle with Matplotlib")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# 8. Show the final plot
#    This opens the plot window and displays the drawing.
plt.show()
```

````python
import matplotlib.pyplot as plt

class CircleDrawer:
    def __init__(self, radius, color='blue'):
        if radius <= 0:
            raise ValueError("Radius must be positive.")
        self.radius = radius
        self.color = color

    def draw(self):
        """Creates a plot and draws the circle on it."""
        # Get the current axes instance, or create one if it doesn't exist.
        ax = plt.gca()

        # Create a Circle patch
        # (0, 0) is the center, `radius` is its radius, `fc` is facecolor.
        circle_patch = plt.Circle((0, 0), radius=self.radius, fc=self.color)

        # Add the patch to the axes
        ax.add_patch(circle_patch)

        # Set the aspect ratio of the plot to be equal.
        # This is crucial for the circle to not look like an ellipse.
        ax.set_aspect('equal', adjustable='box')
        
        # Set plot limits to make sure the full circle is visible
        plt.xlim(-self.radius * 1.2, self.radius * 1.2)
        plt.ylim(-self.radius * 1.2, self.radius * 1.2)
        
        # Add a title
        plt.title(f"A {self.color.capitalize()} Circle with Radius {self.radius}")

        # Display the plot
        plt.show()

# Create an instance of our drawer and call the draw method
red_circle = CircleDrawer(radius=10, color="red")
red_circle.draw()

green_circle = CircleDrawer(radius=5, color="green")
green_circle.draw()
````
