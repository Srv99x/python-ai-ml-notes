# OOP Advanced: Multiple Inheritance and the Method Resolution Order (MRO)

This notebook delves into multiple inheritance, a powerful but complex feature of Python's object-oriented system. We will explore how a class can inherit from more than one parent and, critically, how Python decides which parent's method to use when there are overlaps.

**Key Learning Objectives:**
1.  **Multiple Inheritance**: Understand the syntax and basic concept of a class inheriting from several parent classes.
2.  **Method Resolution Order (MRO)**: Learn what the MRO is and why it's crucial for predicting behavior in multiple inheritance scenarios.
3.  **The "Diamond Problem"**: See a classic example of the complexities that can arise and how Python's MRO elegantly solves it.
4.  **Best Practices**: Discuss when to use and when to avoid multiple inheritance.
