# OOP Advanced: Private and Protected Attributes

In Python, there isn't a strict "private" or "protected" keyword like in languages such as Java or C++. Instead, Python uses naming conventions to signify that an attribute should not be accessed directly from outside the class. This is a system built on trust and convention, but it also includes a mechanism to prevent accidental access.

**Key Learning Objectives:**
1.  **Protected Attributes (`_single_underscore`)**: Understand the convention for attributes intended for internal use by subclasses.
2.  **Private Attributes (`__double_underscore`)**: Learn how Python uses "name mangling" to make attributes harder to access from outside.
3.  **Name Mangling**: See how the interpreter changes the name of private attributes to avoid conflicts.
4.  **Encapsulation**: Understand the purpose of these conventions—to hide implementation details and protect data from accidental modification.

### How "Private" Attributes Work in Python

Unlike languages like Java or C++, Python doesn't have true `private` keywords to strictly enforce data hiding. Instead, it relies on a convention.

To indicate that an attribute is intended for internal use and not part of the public API of the class, you prefix its name with a **double underscore (`__`)**. This is a signal to other developers that they should not access this attribute directly from outside the class.

This triggers a mechanism called **name mangling**. Python automatically changes the name of the attribute to `_ClassName__attributeName`. This makes it much harder to access accidentally from outside the class, but not impossible.

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        # We provide a public method to access the balance
        print("Getting balance via public method...")
        return self.__balance

# Create an instance
acc = BankAccount("John Doe", 1000)

# Accessing public attribute works fine
print(f"Account Owner: {acc.owner}")

# Try to access the private attribute directly (this will fail)
try:
    print(f"Trying to access private balance directly: {acc.__balance}")
except AttributeError as e:
    print(f"Error: {e}")

print("-" * 20)

# Use the public method to get the balance
current_balance = acc.get_balance()
print(f"Current Balance: ${current_balance}")

# Interact with the private attribute via public methods
acc.deposit(500)
acc.withdraw(200)
acc.withdraw(2000) # Invalid
````

### Protected Attributes: The Single Underscore

A **single underscore (`_`)** prefix is another convention. It's used to signal that an attribute or method is "protected". This means it's not intended to be used from outside the class, but it *is* okay for subclasses to access and modify it.

Unlike the double underscore, the single underscore **does not trigger name mangling**. It's purely a hint for the developer.

```python
class Base:
    def __init__(self):
        self._protected_var = 10

class Derived(Base):
    def __init__(self):
        super().__init__()
        print("Accessing protected var from subclass")
        self._protected_var += 5
        print(f"New value: {self._protected_var}")

d = Derived()
# You can still access it, but the underscore signals you shouldn't
print(f"Accessing from outside: {d._protected_var}")
```

### Common Mistakes & Takeaways

1.  **Thinking Attributes are Truly Private**: Remember that name mangling is not a security feature. A determined developer can still access the mangled name (`_ClassName__attributeName`). The goal is to prevent accidental modification, not malicious intent.
2.  **Overusing Private Attributes**: Don't make everything private by default. Start with public attributes and make them private only when you have a good reason—for example, if changing an attribute's value requires validation or triggers some other logic (like in our `BankAccount` example).
3.  **Ignoring the Conventions**: When you see an attribute starting with `_` or `__`, respect the original developer's intent. Avoid accessing them directly unless you have a very good reason and understand the implications.

**Takeaway**: Use `__double_underscore` for attributes that should not be changed directly and should not be overridden by subclasses. Use `_single_underscore` for internal helper attributes that subclasses might need to interact with.
