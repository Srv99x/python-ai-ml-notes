# OOP Advanced: Accessing Private Attributes with Properties

This notebook focuses on the concept of **encapsulation**—bundling data (attributes) and the methods that operate on that data within a single unit (a class). A key part of encapsulation is controlling access to that data.

While Python's "private" attributes (`__`) prevent accidental direct access, we still need a controlled way to read or modify them. This is done using **properties**, which allow you to expose methods as if they were attributes.

**Key Learning Objectives:**
1.  **The Problem**: Understand why direct, uncontrolled access to attributes can be problematic.
2.  **The "Java" Way**: See how getters and setters are used for controlled access.
3.  **The "Pythonic" Way**: Learn to use the `@property` decorator to create clean, readable getters.
4.  **Property Setters and Deleters**: Implement `@<property_name>.setter` and `@<property_name>.deleter` to control modification and deletion.

### Why Access Private Attributes?

First, a word of caution: The primary reason for making an attribute "private" (with a double underscore) is to signal that it is **not meant to be accessed or modified from outside the class**. You should always prefer using public methods (like getters and setters) if the class provides them.

However, there are some legitimate, albeit rare, situations where you might need to access them, such as in debugging, testing, or when working with a class that doesn't provide the necessary public methods.

### The Name Mangling Mechanism

As we learned previously, Python doesn't block access; it just makes it less convenient through a process called **name mangling**. When you create an attribute like `__my_attribute` inside a class named `MyClass`, Python secretly renames it to `_MyClass__my_attribute`.

Knowing this pattern is the key to accessing the attribute directly.

````python
class MyClass:
    def __init__(self):
        self.__private_var = "I am a secret."

# Create an instance
obj = MyClass()

# This will fail because of name mangling
try:
    print(obj.__private_var)
except AttributeError as e:
    print(f"Direct access failed: {e}")

# This is how you can access it if you absolutely must
# The format is _ClassName__attributeName
print(f"Access via name mangling: {obj._MyClass__private_var}")

# You can even modify it (but this is generally a bad idea!)
obj._MyClass__private_var = "The secret is out!"
print(f"Modified value: {obj._MyClass__private_var}")
````

### The "Pythonic" Way: Using Properties

Directly accessing mangled names is clumsy and breaks encapsulation. The proper, "Pythonic" way to provide managed access to private attributes is by using **properties**.

A property is a special kind of attribute that intercepts access (getting, setting, deleting) and redirects it to methods. This allows you to run code and apply logic whenever an attribute is accessed, while still providing a clean, simple syntax to the user of the class.

The `@property` decorator is used to create a "getter" method.
The `@attribute.setter` decorator is used to create a "setter" method.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary # Private

    @property
    def salary(self):
        """
        This is the 'getter' method for salary.
        It's called when you do `emp.salary`.
        """
        print("(Getting salary...)")
        # You could add logic here, like checking permissions
        return self.__salary

    @salary.setter
    def salary(self, new_salary):
        """
        This is the 'setter' method.
        It's called when you do `emp.salary = 55000`.
        """
        print("(Setting salary...)")
        if new_salary > 0:
            self.__salary = new_salary
        else:
            print("Salary must be positive!")

# Create an instance
emp = Employee("Jane Doe", 50000)

# Access the salary via the property (this calls the getter)
print(f"Initial Salary: ${emp.salary}")

# Change the salary via the property (this calls the setter)
emp.salary = 55000
print(f"New Salary: ${emp.salary}")

# Try to set an invalid salary
emp.salary = -100
print(f"Final Salary: ${emp.salary}")
```

### Takeaways

1.  **Avoid Direct Access**: Accessing mangled names (`_ClassName__attributeName`) should be your last resort, reserved for debugging or testing.
2.  **Properties are Preferred**: Properties are the idiomatic Python solution for providing managed access to attributes. They combine the simple syntax of direct attribute access with the power and control of methods.
3.  **Encapsulation**: Using properties allows the class to maintain control over its internal state. The internal `__salary` attribute can be managed and validated by the property's logic, preventing the user of the class from setting it to an invalid state.
