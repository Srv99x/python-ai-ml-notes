Abstraction is a core principle of object-oriented programming that focuses on hiding complex implementation details while exposing only the essential features of an object. It's about creating a simplified, high-level interface for a more complex underlying system.

In Python, abstraction is formally achieved using **Abstract Base Classes (ABCs)**, provided by the `abc` module.

### What is Abstraction?

Abstraction is the concept of hiding complex implementation details and showing only the essential features of an object. It helps in managing complexity by allowing us to focus on what an object does instead of how it does it.

In Python, abstraction is typically achieved using **Abstract Base Classes (ABCs)**. An ABC is a class that cannot be instantiated on its own and is meant to be subclassed. It can define a set of methods and properties that its subclasses are required to implement.

The `abc` module in Python provides the tools for creating ABCs.
- `ABC`: A helper class to create an ABC.
- `@abstractmethod`: A decorator to define an abstract method in an ABC. Subclasses must implement this method.

```python
from abc import ABC, abstractmethod

# Define an Abstract Base Class
class Vehicle(ABC):
    def __init__(self, brand):
        self.brand = brand

    @abstractmethod
    def start_engine(self):
        # This is an abstract method. It has no implementation here.
        # Subclasses MUST provide their own implementation.
        pass

    @abstractmethod
    def stop_engine(self):
        pass

    def get_brand(self):
        # A concrete method can also exist in an ABC
        return f"Brand: {self.brand}"

# Create a subclass that implements the abstract methods
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started. Vroom vroom!"

    def stop_engine(self):
        return "Car engine stopped."

# Another subclass
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started. Rrrrumble!"

    def stop_engine(self):
        return "Motorcycle engine stopped."

# Try to instantiate the abstract class itself (this will fail)
try:
    my_vehicle = Vehicle("Generic")
except TypeError as e:
    print(f"Error: {e}")

print("-" * 20)

# Instantiate the concrete subclasses
my_car = Car("Toyota")
my_motorcycle = Motorcycle("Harley")

print(my_car.get_brand())
print(my_car.start_engine())
print(my_car.stop_engine())
print("-" * 20)
print(my_motorcycle.get_brand())
print(my_motorcycle.start_engine())
print(my_motorcycle.stop_engine())
```

### Common Mistakes & Takeaways

1.  **Forgetting to Implement All Abstract Methods**: If you create a subclass from an ABC but fail to implement even one of its abstract methods, you will get a `TypeError` when you try to instantiate that subclass.
2.  **Instantiating the Abstract Class**: The whole point of an ABC is to serve as a blueprint, not to be used directly. Attempting to create an object from an ABC will always result in a `TypeError`.
3.  **Overusing Abstraction**: Abstraction is a powerful tool, but not every situation requires it. For simple class hierarchies, creating ABCs can be overkill. Use them when you need to enforce a strict contract or API for a group of related classes.

**Takeaway**: Abstraction helps you define a clear and required structure for your subclasses. It ensures that all concrete implementations will have a consistent interface, which leads to more predictable and maintainable code. It's a way of saying, "any class that is a `Vehicle` *must* know how to `start_engine` and `stop_engine`."
