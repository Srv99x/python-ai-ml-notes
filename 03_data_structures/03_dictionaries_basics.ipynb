## Dictionary BasicsThis notebook covers the basics of dictionaries in Python, including creation, access, update, and iteration.It provides a comprehensive overview of how dictionaries work and how to use them effectively.

### Dictionary CreationDictionaries are used to store data values in key-value pairs. They are unordered, mutable, and do not allow duplicate keys.The syntax for creating a dictionary is: `dict = {key: value}`

```python
# Create a dictionary with various data typesinfo = {    # "key" : "value"    "name" : "sourav",    "age" : 19,    "marks" : [90, 99, 98],    "is_adult" : True,    "subjects" : ["py", "Cpp", "html", "js" ],    "topics" : ("OOPs", "loopps", "conditions"),    "CGPA" : 9.8}
```

### Accessing Dictionary ValuesDictionary values can be accessed using their corresponding keys.The output will be the value associated with the key.Common mistake: Trying to access a key that does not exist in the dictionary.Quick tip: Use the `get()` method to avoid KeyError.

### Updating Dictionary ValuesDictionary values can be updated using their corresponding keys.If the key does not exist, a new key-value pair will be added.The syntax for updating a dictionary is: `dict[key] = value`

```python
# Update a dictionary valueinfo["name"] = "srv"  # overwriteinfo["surname"] = "chakraborty"
```

### Creating an Empty DictionaryAn empty dictionary can be created using the `dict()` constructor or the `{}` syntax.The syntax for creating an empty dictionary is: `dict = {}` or `dict = dict()`

```python
# Create an empty dictionarynull_dict = {}null_dict["name"] = "sourav"
```

### Nested DictionariesDictionaries can be nested inside each other.The syntax for creating a nested dictionary is: `dict = {key: {key: value}}`

```python
# Create a nested dictionarystudent = {    "name" : "sourav",    "subMarks" : {        "maths" : 96,        "chem" : 99,        "phy" : 98    }}
```

### Dictionary MethodsDictionaries have several built-in methods, including `keys()`, `values()`, and `len()`.The `keys()` method returns a view object that displays a list of all the available keys.The `values()` method returns a view object that displays a list of all the available values.The `len()` method returns the number of items in the dictionary.

```python
# Use dictionary methodsprint(list(student.keys()))print(len(student))print(list(student.values()))
```

## Key Takeaways* Dictionaries are used to store data values in key-value pairs.* Dictionaries are unordered, mutable, and do not allow duplicate keys.* Dictionary values can be accessed using their corresponding keys.* Dictionary values can be updated using their corresponding keys.* An empty dictionary can be created using the `dict()` constructor or the `{}` syntax.* Dictionaries can be nested inside each other.* Dictionaries have several built-in methods, including `keys()`, `values()`, and `len()`.
