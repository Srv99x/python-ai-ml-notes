## Set OperationsThis notebook covers set operations such as union, intersection, and difference.It provides examples and explanations to help understand these concepts.

### Set UnionThe union of two sets is a set containing all elements that are in either set.Syntax: `set1.union(set2)`

```python
s1 = {1, 4, 5, 8}  # define set 1s2 = {4, 9, 6, 10}  # define set 2print(s1.union(s2))  # print the union of s1 and s2
```

The output of the union operation is a set containing all unique elements from both sets.Common mistake: forgetting that sets only contain unique elements.Quick tip: use the union method to combine sets and remove duplicates.

### Set IntersectionThe intersection of two sets is a set containing all elements that are in both sets.Syntax: `set1.intersection(set2)`

```python
print(s1.intersection(s2))  # print the intersection of s1 and s2
```

The output of the intersection operation is a set containing all elements that are common to both sets.Common mistake: expecting the intersection to contain all elements from both sets.Quick tip: use the intersection method to find common elements between sets.

## Key Takeaways* Set union: combines all unique elements from two sets using `set1.union(set2)`* Set intersection: finds common elements between two sets using `set1.intersection(set2)`
