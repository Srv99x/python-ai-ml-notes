## Set Basics: Uniqueness, Add/Remove, Membership
This notebook covers the basics of sets in Python, including uniqueness, adding and removing elements, and membership testing.
It also discusses the unordered and unindexed nature of sets.

### Creating Sets
Sets are created using the `set()` function or the `{}` syntax. They are unordered collections of unique elements.
The `add()` method is used to add elements to a set.

```python
s = set()  # Create an empty set
s.add(5)   # Add an element to the set
s.add(1)   # Add another element to the set
s.add(10)  # Add another element to the set
s.add(3)   # Add another element to the set
print(s)   # Print the set
```

### Set Output
The output of a set is a collection of unique elements, but the order is not guaranteed.
A common mistake is to assume that sets maintain the order of insertion.
A quick tip is to use sets when you need to store a collection of unique elements, but don't care about the order.

### Creating Sets with Initial Elements
Sets can be created with initial elements using the `{}` syntax.
The `remove()` method is used to remove elements from a set.

```python
s2 = {10, 3, 50, 1, 7, 2, 99, 2}  # Create a set with initial elements
print(s2)  # Print the set
s2.remove(2)  # Remove an element from the set
print(s2)  # Print the set again
```

### Set Behavior
Sets do not store elements in insertion order, but instead use hashing to store elements.
The order of elements in a set can vary depending on the Python version, memory layout, and internal optimizations.
A common mistake is to assume that sets maintain a consistent order across different Python versions.
A quick tip is to use sets when you need to store a collection of unique elements, but don't care about the order.

## Key Takeaways
* Sets are unordered collections of unique elements.
* Sets are created using the `set()` function or the `{}` syntax.
* The `add()` method is used to add elements to a set.
* The `remove()` method is used to remove elements from a set.
* Sets do not maintain the order of insertion.
* Sets use hashing to store elements.
