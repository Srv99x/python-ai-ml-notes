## Modules and Imports in PythonThis topic covers the basics of importing and using modules in Python.We will explore how to import modules and use their functions.

### Importing ModulesIn Python, we use the `import` statement to import modules.The `pyttsx3` module is used for text-to-speech conversion.

```python
# Import the pyttsx3 moduleimport pyttsx3# Initialize the engineengine = pyttsx3.init()# For Mac, If you face error related to 'pyobjc' when running the `init()` method :# Install 9.0.1 version of pyobjc : 'pip install pyobjc>=9.0.1'# Use the engine to say a phraseengine.say('Dibakar rendi')# Run and wait for the speech to finishengine.runAndWait()
```

### Understanding the OutputThe output of this code will be the speech synthesized by the `pyttsx3` module.One common mistake is not installing the required version of `pyobjc` on Mac.A quick tip is to use the `say()` method to change the speech output.

## Key Takeaways* Importing modules using the `import` statement* Initializing the `pyttsx3` engine* Using the `say()` method for text-to-speech conversion* Running and waiting for the speech to finish using `runAndWait()`
