## Recursion Basics: Understanding Base Case and Recursive Case Design
This notebook covers the fundamental concept of recursion in Python, focusing on designing base cases and recursive cases. It provides a step-by-step guide on how to implement recursive functions, using the factorial function as an example.
By the end of this notebook, you will understand how to apply recursion to solve problems and analyze the output of recursive functions.

### What is Recursion?
Recursion is a programming technique where a function calls itself repeatedly until it reaches a base case that stops the recursion. The recursive case is where the function calls itself, and the base case is the condition that stops the recursion.
The syntax pattern for recursion involves defining a function with a base case and a recursive case.

```python
def factorial(n):
    # Base case: if n is 0 or 1, return 1
    if(n==0 or n==1):
        return 1
    # Recursive case: n * factorial(n-1)
    return n*factorial(n-1)
```

### Understanding the Output
The output of the factorial function is the result of the recursive calls. For example, if we call `factorial(4)`, the output will be `4 * 3 * 2 * 1 = 24`.
A common mistake is not defining a proper base case, which can lead to infinite recursion.
A quick tip is to use print statements or a debugger to visualize the recursive calls and understand the flow of the function.

### Testing the Function
To test the function, we need to input a number and print the result.
The syntax pattern for input and output involves using the `input()` function to get user input and the `print()` function to display the result.

```python
n = int(input("Enter a number to find it's factorial:"))
print(f"The factorial is: {{factorial(n)}}".format(factorial(n)))
```

### Output Explanation
The output of the code will be the factorial of the input number. For example, if we input `5`, the output will be `The factorial is: 120`.
A common mistake is not handling invalid inputs, such as negative numbers or non-integer values.
A quick tip is to add error handling to ensure the function works correctly for all possible inputs.
