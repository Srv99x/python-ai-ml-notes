# Loops (`for` and `while`)

### Concept
Loops allow you to execute a block of code repeatedly.

1.  **`for` loops**: Best when you know exactly how many times you want to iterate or when you are iterating over a sequence (like a list, string, or range).
2.  **`while` loops**: Best when you want to keep repeating a block of code *as long as* a certain condition is True. You might not know how many times it will run beforehand.

````python
# The `for` loop
# Use exactly when you know you want to iterate over a specific sequence.

# Iterating over a string (which is a sequence of characters)
for char in "Python":
    print(char)

# Iterating using the range() function
# range(start, stop, step). The 'stop' value is exclusive.
for i in range(1, 6):   # Will print 1, 2, 3, 4, 5
    print(f"Number: {i}")
````

### Using `while` loops

The `while` loop checks a condition. If the condition is True, it runs the block of code. After the block finishes, it checks the condition *again*. If True, it repeats. If False, it stops.

You **MUST** ensure that something inside the loop eventually changes the condition to False. Otherwise, you will create an **infinite loop**, locking up your program until you forcefully stop it.

````python
# A while loop counting down
countdown = 5

while countdown > 0:
    print(countdown)
    # CRITICAL: We MUST subtract from countdown or it will loop forever!
    countdown -= 1 

print("Ignition! 🚀")
````

### 🤖 AI/ML Connection: Loops in Model Training

Loops define the core life cycle of Machine Learning models:
*   **The Training Loop**: ML models learn by repeatedly looking at data (Epochs). A typical training cycle uses a generic `for epoch in range(num_epochs):` loop.
*   **Gradient Descent**: In optimization algorithms, weights adjust repeatedly in a `while` loop until error falls below a certain threshold.
*   **Data loading loops**: Models process data in smaller chunks ("batches"). A loop feeds each batch iteratively into the model to avoid Out Of Memory (OOM) errors.

````python
# A simplified ML Loop concept
epochs = 3 
loss = 100

print(f"Starting Training! Initial Loss: {loss}%")

for epoch in range(1, epochs + 1):
    # Imagine this is our Model looking at data and learning...
    print(f"--- Epoch {epoch} ---")
    
    # Simulating the loss dropping as the model "learns"
    loss = loss / 2 
    print(f"Model Loss improved to: {loss}%")

print("Training Complete.")
````

### 💥 Gotcha: Infinite `while` Loops

The biggest pitfall for beginners using `while` is forgetting to change the condition within the loop block.

```python
x = 0
# while x < 5:    # True
    # print(x)    # This will print 0 forever and crash your notebook/editor!
    
    # x MUST have some modifier like `x += 1` here!
```

**✅ Key Takeaway**: Always verify there is an "escape hatch" in your loop's logic that guarantees the condition will eventually evaluate to `False`.

### 📋 Cheat Sheet

*   **Syntax:**
    ```python
    # For loop (Iterates over an iterable)
    for element in sequence:
        # Code executed for every element
    
    # While loop (Iterates until condition breaks)
    while condition_is_true:
        # Code executed repeatedly    
    ```
*   `range(stop)`: Produces 0 to `stop - 1`.   
*   `range(start, stop)`: Produces `start` to `stop - 1`.
*   `range(start, stop, step)`: Uses a step interval.
