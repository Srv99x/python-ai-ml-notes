# Conditionals (`if`, `elif`, `else`)

### Concept
Control flow dictates the order in which statements are executed. Conditionals allow your program to make decisions and execute specific blocks of code only if certain criteria (conditions) are met. An `if` statement evaluates a boolean expression to either `True` or `False`.

Python uses indentation to define code blocks (unlike languages that use `{}`).

*   **`if`**: The initial condition. Code block runs if `True`.
*   **`elif`** (Else If): Checks another condition if the previous `if`/`elif` was `False`. You can have many `elif`s.
*   **`else`**: The fallback. Runs if all previous conditions were `False`.

````python
# Basic if, elif, else example
temperature = 28

if temperature > 30:
    print("It's a hot day! Stay hydrated. ☀️")
elif temperature >= 20:   # Evaluates only if the above `if` is False
    print("It's a beautiful day! Perfect for a walk. 🌤️")
elif temperature > 10:
    print("It's a bit chilly. Bring a light jacket. 🧥")
else:                     # Runs if ALL above conditions are False
    print("It's cold! Bundle up. 🥶")

# Output for 28 will be: "It's a beautiful day! Perfect for a walk. 🌤️"
````

### Logical Operators (`and`, `or`, `not`)

*   `and`: True if **both** operands are True.
*   `or`: True if **at least one** operand is True.
*   `not`: Reverses the boolean value (True becomes False, False becomes True).

````python
# Combining conditions
age = 25
has_license = True
has_car = False

# Using `and` -> Both must be True
if age >= 18 and has_license:
    print("You can legally drive.")
else:
    print("You cannot legally drive.")

# Using `or` -> At least one must be True
if has_car or has_license:
    print("You have some means of transportation (or at least the right to).")

# Using `not` -> Reverses the result
is_raining = False
if not is_raining:
    print("It's clear outside.")
````

### The Ternary Operator (Conditional Expressions)

A compact, one-line `if-else` statement. Great for simple variable assignments based on a condition.
`[value_if_true] if [condition] else [value_if_false]`

````python
# Standard if-else assignment
score = 85
if score >= 60:
    status = "Pass"
else:
    status = "Fail"
print(status)

# The Ternary equivalent
score = 85
status = "Pass" if score >= 60 else "Fail"
print(status)
````

### 🤖 AI/ML Connection: Why Conditionals Matter in Data Science
In ML/AI, conditionals are used everywhere:
- **Data Preprocessing**: Handling missing values depending on their presence (`if pd.isna(value): ...`).
- **Feature Engineering**: Creating new features based on logical combined conditions.
- **Model Logic**: Deciding on models based on input metadata, hyperparameter tuning stops, early stopping, deciding thresholds, etc.

````python
# A simple AI/ML-like logic example
import random
confidence_score = random.random()  # Returns a float between 0.0 and 1.0

if confidence_score >= 0.9:
    prediction = "Cat"
elif confidence_score >= 0.7:
    prediction = "Unsure, possibly Cat"
else:
    prediction = "Not a Cat"

print(f"Confidence: {confidence_score:.2f} -> Prediction: {prediction}")
````

### 💥 Gotcha: Using `=` instead of `==`

The most common mistake!
*   `=`: **Assignment** operator (assigns a value to a variable).
*   `==`: **Equality** operator (checks if two values are equal).

```python
# WRONG
# if x = 5:  # This will throw a SyntaxError

# RIGHT
# if x == 5:
```

### 📋 Cheat Sheet

*   **Syntax:**
    ```python
    if condition1:
        # Runs if condition1 is True
    elif condition2:
        # Runs if condition1 is False and condition2 is True
    else:
        # Runs if both are False
    ```
*   **Comparison Operators**: `==` (equal), `!=` (not equal), `>` (greater), `<` (less), `>=` (greater or equal), `<=` (less or equal).
*   **Logical**: `and`, `or`, `not`
*   **Ternary**: `TrueValue if Condition else FalseValue`
