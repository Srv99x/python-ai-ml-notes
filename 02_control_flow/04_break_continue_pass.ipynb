## Control Flow Keywords: break, continue, and pass
This topic covers the control flow keywords in Python, specifically break, continue, and pass. These keywords are used to control the flow of loops and conditional statements in Python. By the end of this notebook, you will understand how to use these keywords to manipulate the flow of your programs.

### Break Keyword
The break keyword is used to exit a loop prematurely. It is typically used in conjunction with a conditional statement to exit the loop when a certain condition is met. The syntax pattern for the break keyword is: `if condition: break`.

```python
for i in range(50):
    if(i<=25):
        break  # it exits the loop
    print(i)
```

The output of this code will be empty because the loop exits as soon as `i` is less than or equal to 25. A common mistake when using the break keyword is to forget to include a conditional statement, which can cause the loop to exit prematurely. A quick tip is to use the break keyword in conjunction with a conditional statement to exit the loop when a certain condition is met.

### Continue Keyword
The continue keyword is used to skip the current iteration of a loop and move on to the next iteration. It is typically used in conjunction with a conditional statement to skip the current iteration when a certain condition is met. The syntax pattern for the continue keyword is: `if condition: continue`.

```python
for i in range(30):
    if(i==15):
        continue  # skips the iteration
    print(i)
```

The output of this code will be all numbers from 0 to 29, except for 15, because the continue keyword skips the iteration when `i` is equal to 15. A common mistake when using the continue keyword is to forget to include a conditional statement, which can cause the loop to skip iterations unnecessarily. A quick tip is to use the continue keyword in conjunction with a conditional statement to skip the current iteration when a certain condition is met.

### Pass Keyword
The pass keyword is a placeholder statement that does nothing. It is typically used as a placeholder when a statement is required syntactically, but no execution of code is necessary. The syntax pattern for the pass keyword is: `pass`.

```python
for i in range(10):
    pass
```

The output of this code will be empty because the pass keyword does nothing. A common mistake when using the pass keyword is to forget that it does nothing, which can cause unexpected behavior in the program. A quick tip is to use the pass keyword as a placeholder when a statement is required syntactically, but no execution of code is necessary.

### Example Use Case
Here is an example use case for the pass keyword:

```python
for i in range(11):
    print(i)
```

The output of this code will be all numbers from 0 to 10. This is a simple example of a loop that prints numbers.

## Key Takeaways
* The break keyword is used to exit a loop prematurely.
* The continue keyword is used to skip the current iteration of a loop.
* The pass keyword is a placeholder statement that does nothing.
* These keywords are used to control the flow of loops and conditional statements in Python.
