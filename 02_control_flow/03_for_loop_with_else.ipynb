# The `for...else` Construct

### Concept

This is a uniquely Pythonic feature! In Python, loops can have an `else` block attached to them. 

**The Golden Rule**: The `else` block executes **ONLY IF the loop finishes its execution naturally** (without hitting a `break` statement). 

If the loop is stopped midway via `break`, the `else` block is *skipped*. This is often used for search operations to determine if you exhausted the list without finding your target.

```python
# Scenario 1: Loop finishes perfectly (Hits execution naturally)
print("Scenario 1: Searching for 'Zebra'")
animals = ["Dog", "Cat", "Lion"]

for animal in animals:
    if animal == "Zebra":
        print("Found a Zebra!")
        break
else:
    print("Loop finished naturally -> Did not find a Zebra anywhere!")
    
    
# Scenario 2: Loop hits a 'break' (Interrupts execution)
print("\nScenario 2: Searching for 'Cat'")
for animal in animals:
    if animal == "Cat":
        print("Found a Cat!")
        break  # This stops the loop, AND prevents the `else` from running.
else:
    print("Loop finished naturally -> Did not find a Cat anywhere!")
```
