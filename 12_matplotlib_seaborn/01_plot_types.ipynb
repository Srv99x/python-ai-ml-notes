## Matplotlib Plot TypesThis notebook covers various plot types in Matplotlib, including line plots, scatter plots, bar plots, histograms, box plots, and pie charts.These visualizations are essential for Exploratory Data Analysis (EDA) in AI and ML.

### Line PlotA line plot is used to show trends over time or relationships between continuous variables.The syntax pattern is: `plt.plot(x, y)`.

```python
import matplotlib.pyplot as pltimport numpy as npnp.random.seed(42)x = np.linspace(0, 10, 50)y = 2 * x + np.random.randn(50) * 5plt.figure(figsize=(10, 5))plt.plot(x, y, marker='o', linestyle='-', linewidth=2, markersize=6)plt.xlabel('X')plt.ylabel('Y')plt.title('Line Plot: Trend Over Time')plt.grid(True, alpha=0.3)plt.savefig('plot_01_line.png')plt.close()print("Line plot saved")
```

The output of a line plot shows the trend of the data over time or the relationship between two continuous variables.A common mistake is not labeling the axes or adding a title.A quick tip is to use different markers and line styles to distinguish between multiple lines.

### Multiple Line PlotsMultiple line plots are used to compare trends across multiple series.The syntax pattern is: `plt.plot(x, y1); plt.plot(x, y2)`.

```python
y2 = 1.5 * x + np.random.randn(50) * 4plt.figure(figsize=(10, 5))plt.plot(x, y, label='Series 1', marker='o')plt.plot(x, y2, label='Series 2', marker='s')plt.legend()plt.xlabel('X')plt.ylabel('Y')plt.title('Multiple Line Plots')plt.savefig('plot_02_multi_line.png')plt.close()print("Multiple line plot saved")
```

The output of multiple line plots shows the comparison of trends across multiple series.A common mistake is not using a legend to distinguish between the lines.A quick tip is to use different colors and markers to make the plot more readable.

### Scatter PlotA scatter plot is used to visualize the relationship between two variables.The syntax pattern is: `plt.scatter(x, y)`.

```python
plt.figure(figsize=(10, 5))plt.scatter(x, y, alpha=0.6, s=50)plt.xlabel('Feature 1')plt.ylabel('Feature 2')plt.title('Scatter Plot: Relationship')plt.grid(True, alpha=0.3)plt.savefig('plot_03_scatter.png')plt.close()print("Scatter plot saved")
```

The output of a scatter plot shows the relationship between two variables.A common mistake is not using a sufficient number of data points.A quick tip is to use different colors and markers to distinguish between different groups.

### Colored Scatter PlotA colored scatter plot is used to show a third dimension via color.The syntax pattern is: `plt.scatter(x, y, c=colors)`.

```python
categories_bool = np.random.choice(['Class A', 'Class B'], size=len(x))colors = ['red' if c == 'Class A' else 'blue' for c in categories_bool]plt.figure(figsize=(10, 5))plt.scatter(x, y, c=colors, alpha=0.6, s=50)plt.xlabel('Feature 1')plt.ylabel('Feature 2')plt.title('Scatter Plot: Colored by Class')plt.legend(['Class A', 'Class B'])plt.savefig('plot_04_scatter_colored.png')plt.close()print("Colored scatter plot saved")
```

The output of a colored scatter plot shows the relationship between two variables with a third dimension.A common mistake is not using a legend to distinguish between the colors.A quick tip is to use a color map to show the relationship between the colors and the values.

### Bar PlotA bar plot is used to compare categorical values.The syntax pattern is: `plt.bar(x, y)`.

```python
categories = ['A', 'B', 'C', 'D']values = [23, 45, 56, 78]plt.figure(figsize=(10, 5))plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])plt.ylabel('Values')plt.title('Bar Plot: Compare Categories')for i, v in enumerate(values):    plt.text(i, v + 1, str(v), ha='center')plt.savefig('plot_05_bar.png')plt.close()print("Bar plot saved")
```

The output of a bar plot shows the comparison of categorical values.A common mistake is not labeling the axes or adding a title.A quick tip is to use different colors to distinguish between the categories.

### Horizontal Bar PlotA horizontal bar plot is used to compare categorical values with long category names.The syntax pattern is: `plt.barh(x, y)`.

```python
plt.figure(figsize=(10, 5))plt.barh(categories, values, color=['red', 'blue', 'green', 'orange'])plt.xlabel('Values')plt.title('Horizontal Bar Plot')plt.savefig('plot_06_bar_h.png')plt.close()print("Horizontal bar plot saved")
```

The output of a horizontal bar plot shows the comparison of categorical values with long category names.A common mistake is not labeling the axes or adding a title.A quick tip is to use different colors to distinguish between the categories.

### HistogramA histogram is used to show the distribution of a single variable.The syntax pattern is: `plt.hist(x, bins)`.

```python
data = np.random.randn(1000) * 10 + 50plt.figure(figsize=(10, 5))plt.hist(data, bins=30, edgecolor='black', alpha=0.7)plt.xlabel('Values')plt.ylabel('Frequency')plt.title('Histogram: Distribution of Data')plt.axvline(np.mean(data), color='red', linestyle='--', label=f'Mean: {np.mean(data):.1f}')plt.axvline(np.median(data), color='green', linestyle='--', label=f'Median: {np.median(data):.1f}')plt.legend()plt.savefig('plot_07_histogram.png')plt.close()print("Histogram saved")
```

The output of a histogram shows the distribution of a single variable.A common mistake is not using a sufficient number of bins.A quick tip is to use a density plot to show the distribution of the data.

### Multiple HistogramsMultiple histograms are used to compare the distributions of different groups.The syntax pattern is: `plt.hist(x1, bins); plt.hist(x2, bins)`.

```python
data1 = np.random.randn(500) * 5 + 50data2 = np.random.randn(500) * 8 + 55plt.figure(figsize=(10, 5))plt.hist(data1, bins=25, alpha=0.5, label='Group 1')plt.hist(data2, bins=25, alpha=0.5, label='Group 2')plt.xlabel('Values')plt.ylabel('Frequency')plt.title('Overlapping Histograms')plt.legend()plt.savefig('plot_08_hist_multi.png')plt.close()print("Multi-histogram saved")
```

The output of multiple histograms shows the comparison of the distributions of different groups.A common mistake is not using a legend to distinguish between the groups.A quick tip is to use different colors and transparency to make the plot more readable.

### Box PlotA box plot is used to show the distribution, quartiles, and outliers of a single variable.The syntax pattern is: `plt.boxplot(x)`.

```python
data_groups = [np.random.randn(100) + 50, np.random.randn(100) + 55, np.random.randn(100) + 48]plt.figure(figsize=(10, 5))plt.boxplot(data_groups, labels=['Group A', 'Group B', 'Group C'])plt.ylabel('Values')plt.title('Box Plot: Compare Distributions')plt.grid(True, alpha=0.3)plt.savefig('plot_09_boxplot.png')plt.close()print("Box plot saved")
```

The output of a box plot shows the distribution, quartiles, and outliers of a single variable.A common mistake is not labeling the axes or adding a title.A quick tip is to use different colors to distinguish between the groups.

### Pie ChartA pie chart is used to show the proportions and parts of a whole.The syntax pattern is: `plt.pie(x)`.

```python
sizes = [30, 25, 20, 25]labels = ['Category A', 'Category B', 'Category C', 'Category D']colors = ['red', 'blue', 'green', 'orange']plt.figure(figsize=(8, 8))plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)plt.title('Pie Chart: Proportions')plt.savefig('plot_10_pie.png')plt.close()print("Pie chart saved")
```

The output of a pie chart shows the proportions and parts of a whole.A common mistake is not using a legend to distinguish between the categories.A quick tip is to use different colors to distinguish between the categories.

### Combination and Text AnnotationCombination and text annotation are used to mix plot types and add clarity to the plot.The syntax pattern is: `plt.plot(x, y); plt.text(x, y, text)`.

```python
plt.figure(figsize=(10, 5))plt.scatter(x, y, alpha=0.6, s=50, label='Data points')z = np.polyfit(x, y, 1)p = np.poly1d(z)plt.plot(x, p(x), 'r-', linewidth=2, label=f'Fit: y={z[0]:.2f}x+{z[1]:.2f}')plt.xlabel('X')plt.ylabel('Y')plt.title('Scatter with Trend Line')plt.legend()plt.grid(True, alpha=0.3)plt.savefig('plot_11_scatter_fit.png')plt.close()print("Scatter with fit line saved")
```

The output of a combination plot shows the relationship between two variables with a trend line.A common mistake is not using a legend to distinguish between the data points and the trend line.A quick tip is to use different colors and markers to distinguish between the data points and the trend line.

## Key Takeaways* Line plots: trends over time or continuous relationships.* Scatter plots: relationships between two variables, cluster detection.* Bar plots: compare values across categorical groups.* Histograms: distribution shapes, skewness, outliers.* Box plots: quartiles and outlier detection across groups.* Pie charts: part-to-whole proportions.* Always label axes, add titles, and use legends for clarity.* Color and markers aid interpretation and help distinguish groups.
