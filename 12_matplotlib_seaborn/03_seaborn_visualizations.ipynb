## Seaborn Visualizations
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. This notebook covers various types of visualizations using seaborn, including scatter plots, regression plots, heatmaps, box plots, violin plots, histograms, bar plots, count plots, pair plots, strip plots, swarm plots, line plots, facet grids, and cluster maps.

### Importing Libraries
To start using seaborn, we need to import the necessary libraries. These include seaborn, matplotlib, numpy, and pandas.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Setting Style
Seaborn provides several styles to customize the appearance of plots. We can set the style using the `set_theme` function.

```python
sns.set_theme(style="whitegrid")
```

### Creating Sample Dataset
To demonstrate various visualizations, we need a sample dataset. We will create a DataFrame with random data.

```python
np.random.seed(42)
data = pd.DataFrame({
    'Age': np.random.randint(20, 60, 200),
    'Salary': np.random.randint(30000, 120000, 200),
    'Department': np.random.choice(['Sales', 'IT', 'HR', 'Marketing'], 200),
    'YearsExp': np.random.randint(0, 30, 200),
    'Performance': np.random.uniform(5, 10, 200)
})
print("Sample dataset loaded")
print(f"Shape: {data.shape}")
```

### RELPLOT: Relationship Plots
Relationship plots are used to visualize the relationship between two variables. We can use the `scatterplot` function to create a scatter plot.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Age', y='Salary', hue='Department', size='Performance', alpha=0.6)
plt.title('Salary vs Age by Department')
plt.savefig('seaborn_scatter.png')
plt.close()
print("Scatter plot saved")
```

The output of the scatter plot shows the relationship between age and salary for each department. The size of the points represents the performance of the employees. One common mistake is not using a suitable marker size, which can make the plot difficult to read. A quick tip is to use the `alpha` parameter to adjust the transparency of the points.

### Regression Plot
A regression plot is used to visualize the linear relationship between two variables. We can use the `regplot` function to create a regression plot.

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='Age', y='Salary', scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('Salary vs Age (with linear fit)')
plt.savefig('seaborn_regplot.png')
plt.close()
print("Regression plot saved")
```

The output of the regression plot shows the linear relationship between age and salary. The red line represents the linear fit. One common mistake is not using a suitable line color, which can make the plot difficult to read. A quick tip is to use the `line_kws` parameter to adjust the line style.

### HEATMAP: Correlation Matrix
A heatmap is used to visualize the correlation between multiple variables. We can use the `heatmap` function to create a heatmap.

```python
plt.figure(figsize=(8, 6))
numeric_data = data[['Age', 'Salary', 'YearsExp', 'Performance']].corr()
sns.heatmap(numeric_data, annot=True, cmap='coolwarm', center=0, 
            cbar_kws={'label': 'Correlation'}, square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('seaborn_heatmap.png')
plt.close()
print("Heatmap saved")
```

The output of the heatmap shows the correlation between the numeric variables. The color bar represents the correlation coefficient. One common mistake is not using a suitable color map, which can make the plot difficult to read. A quick tip is to use the `cmap` parameter to adjust the color map.

### BOXPLOT: Compare Distributions by Group
A box plot is used to compare the distribution of a variable across different groups. We can use the `boxplot` function to create a box plot.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Department', y='Salary', hue='Department', palette='Set2')
plt.title('Salary Distribution by Department')
plt.savefig('seaborn_boxplot.png')
plt.close()
print("Boxplot saved")
```

The output of the box plot shows the distribution of salary across different departments. The box represents the interquartile range, and the whiskers represent the range of the data. One common mistake is not using a suitable palette, which can make the plot difficult to read. A quick tip is to use the `palette` parameter to adjust the color scheme.

### VIOLIN PLOT: Smoother Representation of Distribution
A violin plot is used to visualize the distribution of a variable. We can use the `violinplot` function to create a violin plot.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Department', y='Salary', palette='muted')
plt.title('Salary Distribution (Violin Plot)')
plt.savefig('seaborn_violin.png')
plt.close()
print("Violin plot saved")
```

The output of the violin plot shows the distribution of salary across different departments. The width of the violin represents the density of the data. One common mistake is not using a suitable palette, which can make the plot difficult to read. A quick tip is to use the `palette` parameter to adjust the color scheme.

### HISTPLOT: Distribution with By-Group Comparison
A histogram is used to visualize the distribution of a variable. We can use the `histplot` function to create a histogram.

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=data, x='Salary', hue='Department', kde=True, stat='density', bins=30)
plt.title('Salary Distribution by Department')
plt.savefig('seaborn_histplot.png')
plt.close()
print("Histplot saved")
```

The output of the histogram shows the distribution of salary across different departments. The curve represents the kernel density estimate. One common mistake is not using a suitable number of bins, which can make the plot difficult to read. A quick tip is to use the `bins` parameter to adjust the number of bins.

### BARPLOT: Aggregate Values by Group
A bar plot is used to visualize the aggregate values of a variable across different groups. We can use the `barplot` function to create a bar plot.

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Department', y='Salary', estimator=np.mean, errorbar='sd')
plt.title('Mean Salary by Department (with error bars)')
plt.savefig('seaborn_barplot.png')
plt.close()
print("Barplot saved")
```

The output of the bar plot shows the mean salary across different departments. The error bars represent the standard deviation. One common mistake is not using a suitable estimator, which can make the plot difficult to read. A quick tip is to use the `estimator` parameter to adjust the aggregation function.

### COUNTPLOT: Count Observations by Category
A count plot is used to visualize the count of observations in each category. We can use the `countplot` function to create a count plot.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Department', palette='husl')
plt.title('Employee Count by Department')
plt.savefig('seaborn_countplot.png')
plt.close()
print("Countplot saved")
```

The output of the count plot shows the count of employees in each department. One common mistake is not using a suitable palette, which can make the plot difficult to read. A quick tip is to use the `palette` parameter to adjust the color scheme.

### PAIRPLOT: Matrix of Scatter Plots
A pair plot is used to visualize the relationship between multiple variables. We can use the `pairplot` function to create a pair plot.

```python
data_small = data.sample(50)
sns.pairplot(data_small[['Age', 'Salary', 'YearsExp', 'Performance']], 
             diag_kind='hist', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot: All Relationships', y=1.00)
plt.savefig('seaborn_pairplot.png')
plt.close()
print("Pairplot saved")
```

The output of the pair plot shows the relationship between all pairs of variables. The diagonal plots represent the distribution of each variable. One common mistake is not using a suitable number of samples, which can make the plot difficult to read. A quick tip is to use the `sample` method to adjust the number of samples.

### STRIPPLOT & SWARMPLOT: Categorical Scatter
A strip plot and a swarm plot are used to visualize the distribution of a variable across different categories. We can use the `stripplot` and `swarmplot` functions to create these plots.

```python
plt.figure(figsize=(10, 6))
sns.stripplot(data=data, x='Department', y='Salary', hue='Department', jitter=True, size=8, alpha=0.6)
plt.title('Salary by Department (Strip Plot)')
plt.savefig('seaborn_stripplot.png')
plt.close()
print("Stripplot saved")
```

The output of the strip plot shows the distribution of salary across different departments. The points are jittered to avoid overplotting. One common mistake is not using a suitable jitter, which can make the plot difficult to read. A quick tip is to use the `jitter` parameter to adjust the amount of jitter.

### LINEPLOT: Time Series or Sequence
A line plot is used to visualize the trend of a variable over time or across a sequence. We can use the `lineplot` function to create a line plot.

```python
ts_data = pd.DataFrame({
    'Month': np.tile(range(12), 3),
    'Sales': np.concatenate([
        100 + np.cumsum(np.random.randn(12) * 5),
        110 + np.cumsum(np.random.randn(12) * 5),
        105 + np.cumsum(np.random.randn(12) * 5)
    ]),
    'Region': np.repeat(['North', 'South', 'East'], 12)
})
plt.figure(figsize=(12, 6))
sns.lineplot(data=ts_data, x='Month', y='Sales', hue='Region', marker='o', linewidth=2)
plt.title('Sales Trend by Region')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.savefig('seaborn_lineplot.png')
plt.close()
print("Lineplot saved")
```

The output of the line plot shows the trend of sales over time for each region. The lines are colored according to the region. One common mistake is not using a suitable line style, which can make the plot difficult to read. A quick tip is to use the `linewidth` parameter to adjust the line width.

### FACETGRID: Multiple Subplots by Category
A facet grid is used to create multiple subplots, one for each category. We can use the `FacetGrid` function to create a facet grid.

```python
g = sns.FacetGrid(data, col='Department', col_wrap=2, height=4)
g.map(sns.scatterplot, 'Age', 'Salary', alpha=0.6)
g.set_axis_labels('Age', 'Salary')
plt.savefig('seaborn_facetgrid.png')
plt.close()
print("FacetGrid saved")
```

The output of the facet grid shows the relationship between age and salary for each department. The subplots are arranged in a grid. One common mistake is not using a suitable number of columns, which can make the plot difficult to read. A quick tip is to use the `col_wrap` parameter to adjust the number of columns.

### CLUSTERMAP: Hierarchical Clustering Heatmap
A cluster map is used to visualize the hierarchical clustering of a dataset. We can use the `clustermap` function to create a cluster map.

```python
cluster_data = data[['Age', 'Salary', 'YearsExp', 'Performance']].sample(30).reset_index(drop=True)
cluster_data_normalized = (cluster_data - cluster_data.mean()) / cluster_data.std()
g = sns.clustermap(cluster_data_normalized, cmap='vlag', center=0, method='average')
g.ax_heatmap.set_title('Hierarchical Clustering Heatmap')
plt.savefig('seaborn_clustermap.png')
plt.close()
print("Clustermap saved")
```

The output of the cluster map shows the hierarchical clustering of the dataset. The dendrogram represents the clustering structure. One common mistake is not using a suitable normalization, which can make the plot difficult to read. A quick tip is to use the `normalize` parameter to adjust the normalization.

## Key Takeaways
* Seaborn integrates with pandas DataFrames; use 'data', 'x', 'y', 'hue' parameters.
* Use hue to add categorical dimension; use col/row to create subplots.
* Heatmap: correlation analysis; pairplot: explore all relationships.
* Boxplot/violin: compare distributions across groups; stripplot/swarmplot: show individual points.
* FacetGrid: create multiple plots by category for systematic comparison.
