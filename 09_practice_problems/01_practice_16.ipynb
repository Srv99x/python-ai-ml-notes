## Practice Problems and Concept Reinforcement
This notebook covers practice problems and concept reinforcement for object-oriented programming in Python, specifically focusing on classes and objects.
It includes a simple example of a Student class to demonstrate key concepts.

### Classes and Objects in Python
A class is a template for creating objects. It defines the structure and behavior of an object, including its properties (data) and methods (functions).
The syntax for defining a class is `class ClassName:`.

```python
class Student:
    def __init__(self, name, marks):
        # Initialize the object with a name and a list of marks
        self.name = name
        self.marks = marks

    def get_sum(self):
        # Calculate the sum of the marks
        sum = 0
        for val in self.marks:
            sum += val
        # Print the average marks
        print(f"Hello {self.name} your average marks is {sum/len(self.marks)}")
```

### Output and Explanation
When we create an object of the Student class and call the `get_sum` method, it calculates and prints the average marks.
A common mistake is to hardcode the length of the marks list instead of using `len(self.marks)`.
A quick tip is to use a list comprehension or the built-in `sum` function to simplify the calculation.

### Creating an Object and Calling a Method
To create an object, we use the class name followed by parentheses containing the required arguments.
We can then call methods on the object using the dot notation.

```python
s1 = Student("Sourav", [78, 89, 97])
s1.get_sum()
```

### Output and Explanation
The output will be the average marks of the student Sourav.
A common mistake is to forget to call the method on the object.
A quick tip is to use a debugger or print statements to understand the flow of the program.

## Key Takeaways
* Classes and objects are fundamental concepts in object-oriented programming.
* A class defines the structure and behavior of an object.
* Methods are functions that belong to a class and can be called on objects.
* The `__init__` method is a special method that initializes an object when it is created.
* The `len` function can be used to get the length of a list.
