### Print Statement
The print statement is used to output text to the screen. The basic syntax is `print('text')`. By default, print adds a newline at the end of its output.

```python
print("a")  # prints 'a' followed by a newline
print("b")  # prints 'b' followed by a newline
print("c", end="")  # prints 'c' without a newline
print("d", end="")  # prints 'd' without a newline
```

The output of the above code will be `abc` on the same line, because we used `end=''` to prevent print from adding a newline. A common mistake is forgetting to include the `end=''` argument when you want to print on the same line. A quick tip is to always consider what you want the output to look like and adjust your print statements accordingly.

## Key Takeaways
* The print statement is used to output text to the screen.
* By default, print adds a newline at the end of its output.
* You can prevent print from adding a newline by using the `end=''` argument.
