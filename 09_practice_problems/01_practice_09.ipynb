## Practice Problems and Concept ReinforcementThis notebook covers practice problems to reinforce concepts in Python.It includes examples of loops and pattern printing.

### Getting User Input and Printing PatternsThis section demonstrates how to get user input and print patterns using loops.The `input()` function is used to get user input, and `for` loops are used to print patterns.

```python
n = int(input("Enter a number: "))for i in range(1, n+1):    # Print spaces before the pattern    print(" "*(n-i), end="")    # Print the pattern    print("*" * (2*i-1), end="")    # Move to the next line    print("")
```

The output of this code will be a pyramid pattern with the number of rows specified by the user.A common mistake is to forget to convert the user input to an integer using `int()`. A quick tip is to use the `end` parameter of the `print()` function to print on the same line.

### Printing Inverted PatternsThis section demonstrates how to print inverted patterns using loops.The `range()` function is used to generate numbers for the loop, and the `print()` function is used to print the pattern.

```python
for i in range(1, n + 1):    # Print spaces before the pattern    print(" " * (i - 1), end="")    # Print the pattern    print("*" * (2 * (n - i) + 1))
```

The output of this code will be an inverted pyramid pattern with the number of rows specified by the user.A common mistake is to forget to adjust the number of spaces and asterisks in the pattern.A quick tip is to use the `range()` function to generate numbers for the loop.
