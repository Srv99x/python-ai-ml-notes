## Practice Problems and Concept Reinforcement
This notebook covers practice problems and concept reinforcement for lists and user input in Python.
It includes examples of how to take user input, append it to a list, and print the list.

### Initializing an Empty List
In Python, we can initialize an empty list using square brackets `[]`. This is useful when we want to store data that will be generated or collected later in our program.
The syntax pattern is: `variable_name = []`

```python
# Initialize an empty list to store movie names
movies = []
```

### Taking User Input
The `input()` function in Python is used to take user input. It returns a string by default.
The syntax pattern is: `variable_name = input('prompt_message')`

```python
# Take user input for three movie names
mov1 = input("Enter name of 1st movie: ")
mov2 = input("Enter name of 2nd movie: ")
mov3 = input("Enter name of 3rd movie: ")
```

### Appending to a List
The `append()` method in Python is used to add an element to the end of a list.
The syntax pattern is: `list_name.append(element)`

```python
# Append the movie names to the list
movies.append(mov1)
movies.append(mov2)
movies.append(mov3)
```

### Printing the List
The `print()` function in Python is used to output the contents of a list.
One common mistake is trying to print a list without using the `print()` function.
A quick tip is to use the `print()` function with a descriptive message to make the output more readable.

```python
# Print the list of movie names
print(movies)
```

## Key Takeaways
* Initializing an empty list using square brackets `[]`.
* Taking user input using the `input()` function.
* Appending to a list using the `append()` method.
* Printing the contents of a list using the `print()` function.
