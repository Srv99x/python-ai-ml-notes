### Defining a Function to Find the Greatest of Three Numbers
A function is a block of code that can be called multiple times from different parts of your program.
The `greatest` function takes three arguments and uses conditional statements to determine which one is the greatest.

```python
def greatest(a, b, c):
    # Check if a is greater than b and c
    if a > b and a > c:
        print(f"{a} is greatest!")
    # Check if b is greater than a and c
    elif b > a and b > c:
        print(f"{b} is greatest!")
    # Check if c is greater than a and b
    elif c > b and c > a:
        print(f"{c} is greatest!")
```

### Understanding the Output and Common Mistakes
The output of the `greatest` function will be a message indicating which number is the greatest.
A common mistake is to forget to use the `elif` statement to check for multiple conditions.
A quick tip is to use the `print` function to test the output of your function with different inputs.

### Getting User Input
The `input` function is used to get user input in Python.
It returns a string, so we need to use the `int` function to convert it to an integer.

```python
a = int(input("Enter the 1st no: "))
b = int(input("Enter the 2nd no: "))
c = int(input("Enter the 3rd no: "))
```

### Calling the Function with User Input
Now that we have the user input, we can call the `greatest` function with these values.
The output will be the message indicating which number is the greatest.
A common mistake is to forget to call the function with the correct arguments.
A quick tip is to use the `print` function to test the output of your function with different inputs.

```python
greatest(a, b, c)
```
