### Finding the Greatest Number Among Four Input Numbers
This concept uses conditional statements to compare four input numbers and determine the greatest among them.
The syntax pattern involves using `if` and `elif` statements to compare each number with the others.

```python
# Get four numbers from the user
a1 = int(input("Enter number 1: "))
a2 = int(input("Enter number 2: "))
a3 = int(input("Enter number 3: "))
a4 = int(input("Enter number 4: "))

# Determine the greatest number
if a1 > a2 and a1 > a3 and a1 > a4:
    print("The greatest number is a1", a1)
elif a2 > a1 and a2 > a3 and a2 > a4:
    print("The greatest number is a2", a2)
elif a3 > a2 and a3 > a1 and a3 > a4:
    print("The greatest number is a3", a3)
elif a4 > a2 and a4 > a3 and a4 > a1:
    print("The greatest number is a4", a4)
```

The output of this code will be the greatest number among the four input numbers.
A common mistake is to use the wrong comparison operator (e.g., `=` instead of `>`).
A quick tip is to use a list to store the numbers and then use the `max()` function to find the greatest number.

### Determining Whether a Student Has Passed or Failed an Exam
This concept uses conditional statements to determine whether a student has passed or failed an exam based on their marks.
The syntax pattern involves using `if` and `elif` statements to compare the student's marks with the passing criteria.

```python
# Get the marks for three subjects from the user
marks1 = int(input("Enter marks for 1st subject: "))
marks2 = int(input("Enter marks for 2nd subject: "))
marks3 = int(input("Enter marks for 3rd subject: "))

# Calculate the total percentage
total_percentage = (100 * (marks1 + marks2 + marks3)) / 300

# Determine whether the student has passed or failed
if total_percentage >= 40 and marks1 >= 33 and marks2 >= 33 and marks3 >= 33:
    print("YOU HAVE PASSED YOUR EXAM")
else:
    print("YOU FAILED. TRY AGAIN IN NEXT LIFE")
```

The output of this code will be a message indicating whether the student has passed or failed the exam.
A common mistake is to forget to check for the minimum marks required in each subject.
A quick tip is to use a dictionary to store the marks for each subject and then use a loop to calculate the total percentage.

## Key Takeaways
* Use `if` and `elif` statements to compare numbers and determine the greatest among them.
* Use conditional statements to determine whether a student has passed or failed an exam based on their marks.
* Calculate the total percentage by summing the marks for all subjects and dividing by the total possible marks.
* Check for the minimum marks required in each subject to determine whether the student has passed or failed.
