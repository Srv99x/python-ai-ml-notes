### Calculating FactorialsA factorial of a number n (denoted as n!) is the product of all positive integers less than or equal to n.The syntax pattern is: `product = 1; for i in range(1, n+1): product *= i`

```python
n = int(input("Enter a number: "))product = 1for i in range(1, n+1):    # multiply the current product by i in each iteration    product = product * iprint(f"The factorial of {n} is: {product}")
```
