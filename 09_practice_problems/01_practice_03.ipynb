## Practice Problems and Concept ReinforcementThis notebook covers practice problems and concept reinforcement for dictionaries in Python.It includes creating and updating a dictionary with user input.

### Creating an Empty DictionaryA dictionary in Python is an unordered collection of key-value pairs.It is created using the `{}` syntax or the `dict()` function.

```python
# Create an empty dictionaryd = {}
```

The output of this cell will be an empty dictionary `{}`.A common mistake is to use the `[]` syntax to create a dictionary, which will result in a list.A quick tip is to use the `dict()` function to create a dictionary from a list of key-value pairs.

### Getting User InputThe `input()` function is used to get user input in Python.It returns a string, which can be stored in a variable.

```python
# Get user input for friend's name and languagename = input("Enter friend's name: ")lang = input("Enter language name: ")
```

The output of this cell will be the user's input, which will be stored in the `name` and `lang` variables.A common mistake is to forget to use the `input()` function, which will result in a syntax error.A quick tip is to use the `strip()` function to remove leading and trailing whitespace from the user's input.

### Updating a DictionaryThe `update()` method is used to update a dictionary with new key-value pairs.It takes a dictionary or a list of key-value pairs as an argument.

```python
# Update the dictionary with the user's inputd.update({name: lang})
```

The output of this cell will be the updated dictionary.A common mistake is to use the `append()` method, which will result in a syntax error.A quick tip is to use the `update()` method to update a dictionary with multiple key-value pairs at once.

### Repeating the ProcessTo add more key-value pairs to the dictionary, we can repeat the process of getting user input and updating the dictionary.

```python
# Repeat the process to add more key-value pairsname = input("Enter friend's name: ")lang = input("Enter language name: ")d.update({name: lang})name = input("Enter friend's name: ")lang = input("Enter language name: ")d.update({name: lang})name = input("Enter friend's name: ")lang = input("Enter language name: ")d.update({name: lang})
```

The output of this cell will be the updated dictionary with multiple key-value pairs.A common mistake is to forget to update the dictionary after getting user input, which will result in lost data.A quick tip is to use a loop to repeat the process of getting user input and updating the dictionary.

### Printing the DictionaryThe `print()` function is used to print the dictionary.It will display the key-value pairs in the dictionary.

```python
# Print the dictionaryprint(d)
```

The output of this cell will be the dictionary with all the key-value pairs.A common mistake is to forget to print the dictionary, which will result in no output.A quick tip is to use the `pprint()` function to print the dictionary in a pretty format.
