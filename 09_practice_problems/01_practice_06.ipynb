### Prime Number CheckerA prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.The syntax pattern for checking prime numbers involves using a for loop to iterate through potential divisors.

```python
n = int(input("Enter a number:"))  # Get user input and convert to integerfor i in range(2, n):  # Iterate from 2 to n-1    if(n % i == 0):  # Check if n is divisible by i        print("Number is not prime")  # If divisible, it's not prime        break  # Exit loop since we found a divisorelse:  # If loop completes without finding a divisor    print("Number is prime")  # Then n is prime
```

### Output and Common MistakesThe output will indicate whether the input number is prime or not.A common mistake is not handling the case where the input number is less than 2, which are not prime by definition.A quick tip is to always test your code with edge cases, such as inputs of 1, 2, and 3.

## Key Takeaways* Prime numbers are natural numbers greater than 1 with no divisors other than 1 and themselves.* The syntax pattern for checking prime numbers involves using a for loop to iterate through potential divisors.* Always test your code with edge cases to ensure correctness.
