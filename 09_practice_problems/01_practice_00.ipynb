## Practice Problems and Concept Reinforcement
This notebook covers practice problems and reinforces key concepts in Python, including user input and string manipulation.
It provides a quick recap of these concepts and allows for experimentation with different inputs and values.

### User Input
The `input()` function in Python is used to take user input. It returns a string by default.
The `str()` function is used to ensure the input is a string, although this is not necessary in this case.

```python
name = str(input("enter your name: "))
print(len(name))  # prints the length of the input string
```

The output of this code is the length of the input string. A common mistake is to forget that the `len()` function counts spaces and punctuation as characters.
A quick tip is to use the `strip()` method to remove leading and trailing whitespace from the input string.

### String Manipulation
The `count()` method in Python is used to count the number of occurrences of a substring in a string.
The syntax pattern is `string.count(substring)`.

```python
stmnt = "hi $iam a $very $good person$"
print(stmnt.count("$"))  # prints the number of occurrences of the substring "$"
```

The output of this code is the number of occurrences of the substring "$" in the string. A common mistake is to forget that the `count()` method is case-sensitive.
A quick tip is to use the `lower()` or `upper()` method to convert the string to lowercase or uppercase before counting occurrences.

## Key Takeaways
* The `input()` function is used to take user input.
* The `len()` function is used to get the length of a string.
* The `count()` method is used to count the number of occurrences of a substring in a string.
