### Defining a Function
A function is a block of code that can be called multiple times from different parts of your program.
The syntax pattern is: `def function_name(parameters):`

```python
def multiply(n):
    # Loop through numbers 1-10 and print the multiplication table for the input number
    for i in range(1, 11):
        print(f"{n} X {i} = {n*i}")
```

### Understanding the Output
The output will be the multiplication table for the input number, from 1 to 10.
A common mistake is to forget to indent the code inside the loop or function.
A quick tip is to use the `range()` function to generate a sequence of numbers for the loop.

### Calling the Function
To use the function, simply call it with the desired input value.
The syntax pattern is: `function_name(input_value)`

```python
multiply(6)
```

### Output Explanation
The output will be the multiplication table for the number 6, from 1 to 10.
A common mistake is to expect the function to return a value, when it actually prints the result directly.
A quick tip is to experiment with different input values to see how the function behaves.

## Key Takeaways
* Functions are reusable blocks of code that take parameters and can be called multiple times.
* Loops can be used to generate sequences of numbers or repeat tasks.
* The `range()` function can be used to generate a sequence of numbers for a loop.
* Functions can print output directly, rather than returning a value.
