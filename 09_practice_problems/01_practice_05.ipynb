### While LoopsA while loop is used to execute a block of code as long as a certain condition is true.The syntax pattern is: `while condition: code_to_execute`

```python
n = int(input("Enter a number: "))i = 1while i < 11:    # print the multiplication table for the input number    print(f"{n} X {i} = {n*i}")    i += 1
```

The output of this code will be the multiplication table for the input number.A common mistake is to forget to increment the loop variable (`i += 1`), causing an infinite loop.A quick tip is to use a while loop when you don't know the number of iterations in advance.

### For LoopsA for loop is used to execute a block of code for each item in a sequence.The syntax pattern is: `for variable in sequence: code_to_execute`

```python
# alternative way to print the multiplication table using a for loop# for i in range(1, 11):#     print(f"{n} X {i} = {n*i}")
```

This code is an alternative way to print the multiplication table using a for loop.A common mistake is to forget to specify the range of the loop (`range(1, 11)`).A quick tip is to use a for loop when you need to iterate over a sequence.
