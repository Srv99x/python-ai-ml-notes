## Pandas Series and DataFrame Creation
This notebook covers the basics of creating and manipulating Pandas Series and DataFrames, essential data structures for data analysis and machine learning in Python.
It introduces how to create Series and DataFrames from various sources, access and manipulate their elements, and perform basic statistical operations.

### SERIES: 1D LABELED DATA
Series are like labeled lists, mapping indices to values. They are used for individual features, time series, or target variables.

```python
import pandas as pd
import numpy as np

# Create Series from list
s1 = pd.Series([10, 20, 30, 40])
print("Series from list:")
print(s1)
print("\nDefault index:", s1.index.tolist())
```

The output shows the Series created from the list with default integer indexing. The `index` attribute of a Series returns its index labels.
A common mistake is trying to access elements by their position using the `[]` operator without specifying the index label. A quick tip is to use the `loc` method for label-based indexing.

### Create Series with Custom Index
Creating a Series with a custom index allows for more meaningful labels, making the data more interpretable.

```python
# Create Series with custom index
s2 = pd.Series([85, 92, 88, 95], index=['Alice', 'Bob', 'Charlie', 'Diana'])
print("\nSeries with custom index:")
print(s2)
```

The output displays the Series with the specified custom index. Accessing elements by their label is more readable and self-documenting than using positional indexing.
A common mistake is forgetting to specify the index when creating a Series from a list. A quick tip is to use the `index` parameter when creating a Series to ensure meaningful labels.

### Access Element by Label
Accessing elements by their label is a key feature of Series, making the code more readable and self-documenting.

```python
# Access element by label
print("\nAlice's score:", s2['Alice'])
print("Bob's score:", s2.loc['Bob'])
```

The output shows how to access elements by their label using both the `[]` operator and the `loc` method. Both methods are label-based and provide a clear, readable way to access data.
A common mistake is mixing label-based and positional indexing, which can lead to confusion. A quick tip is to consistently use label-based indexing for clarity and readability.

### Create Series from Dictionary
Creating a Series from a dictionary automatically uses the dictionary keys as the index labels.

```python
# Create Series from dictionary
s3 = pd.Series({'apple': 5, 'banana': 7, 'orange': 3})
print("\nSeries from dict:")
print(s3)
```

The output displays the Series created from the dictionary, with the dictionary keys used as the index labels. This method is convenient for creating Series from data that is naturally represented as key-value pairs.
A common mistake is forgetting that the dictionary keys become the index labels. A quick tip is to choose meaningful dictionary keys to ensure clear and interpretable index labels.

### Series Properties
Understanding the properties of a Series, such as its data type and length, is essential for working with the data.

```python
# Series properties
print("\nSeries dtype:", s2.dtype)
print("Series length:", len(s2))
print("Series values:", s2.values)  # NumPy array underlying the Series
```

The output shows the data type, length, and underlying NumPy array of the Series. These properties are crucial for understanding the structure and content of the Series.
A common mistake is not checking the data type of the Series, which can lead to errors when performing operations. A quick tip is to always verify the data type of the Series to ensure compatibility with subsequent operations.

## DATAFRAME: 2D LABELED DATA
DataFrames are the most common data structure in machine learning, representing 2D labeled data where each column is a feature and each row is a sample.

```python
# Create DataFrame from dictionary of lists
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 30, 28, 22],
    'Salary': [50000, 60000, 55000, 48000],
    'Department': ['Sales', 'IT', 'HR', 'Sales']
}
df = pd.DataFrame(data_dict)
print("\nDataFrame from dict:")
print(df)
```

The output displays the DataFrame created from the dictionary of lists. This is a standard format for loading datasets, where each key in the dictionary becomes a column in the DataFrame.
A common mistake is not ensuring that all lists in the dictionary are of the same length, which can lead to errors. A quick tip is to verify that all lists have the same length before creating the DataFrame.

### DataFrame Properties
Understanding the properties of a DataFrame, such as its shape, columns, and data types, is crucial for working with the data.

```python
# DataFrame properties
print("\nDataFrame shape:", df.shape)  # (4, 4) → 4 rows, 4 columns
print("DataFrame columns:", df.columns.tolist())
print("DataFrame dtypes:\n", df.dtypes)
print("\nDataFrame info:")
df.info()  # Concise summary of data types and missing values
```

The output shows the shape, columns, data types, and a concise summary of the DataFrame. These properties are essential for understanding the structure and content of the DataFrame.
A common mistake is not checking for missing values, which can affect subsequent analyses. A quick tip is to use the `info` method to get a concise summary of the DataFrame, including information about missing values.

### Access Column as Series
Accessing a column of a DataFrame as a Series allows for operations specific to Series, such as statistical calculations.

```python
# Access column as Series
names = df['Name']
ages = df['Age']
print("\nAccess 'Age' column:", ages.values)
```

The output shows how to access a column as a Series and then access its values. This is useful for performing operations that are specific to Series.
A common mistake is trying to perform Series-specific operations directly on the DataFrame. A quick tip is to access the column as a Series when performing operations that are specific to Series.

### Multiple Columns
Selecting multiple columns from a DataFrame allows for the creation of a new DataFrame with a subset of the original columns.

```python
# Multiple columns
subset = df[['Name', 'Salary']]
print("\nSubset [Name, Salary]:")
print(subset)
```

The output displays the subset of the DataFrame with the specified columns. This is useful for focusing on specific features of the data.
A common mistake is not specifying the correct column names. A quick tip is to verify the column names before selecting them to ensure the correct subset is created.

## CREATE DATAFRAME FROM LISTS OF DICTS
Creating a DataFrame from a list of dictionaries is an alternative format that can be more intuitive for row-based data.

```python
# Create DataFrame from list of dicts
records = [
    {'Name': 'Alice', 'Age': 25, 'Salary': 50000},
    {'Name': 'Bob', 'Age': 30, 'Salary': 60000},
    {'Name': 'Charlie', 'Age': 28, 'Salary': 55000}
]
df2 = pd.DataFrame(records)
print("\nDataFrame from list of dicts:")
print(df2)
```

The output shows the DataFrame created from the list of dictionaries. This format is useful when the data is naturally represented as a list of records.
A common mistake is not ensuring that all dictionaries have the same keys. A quick tip is to verify that all dictionaries have the same keys to ensure the correct DataFrame is created.

## CREATE DATAFRAME FROM NUMPY ARRAYS
Converting NumPy arrays to DataFrames is useful for integrating numerical computations with data analysis.

```python
# Create DataFrame from NumPy array
arr = np.random.randn(3, 3)
df3 = pd.DataFrame(arr, columns=['Feature1', 'Feature2', 'Feature3'])
print("\nDataFrame from NumPy array:")
print(df3)
```

The output displays the DataFrame created from the NumPy array. This is useful for converting numerical data into a format suitable for data analysis.
A common mistake is not specifying the column names. A quick tip is to specify meaningful column names to ensure the DataFrame is interpretable.

## DATAFRAME INDEXING AND LABELS
Using meaningful indices and labels in DataFrames makes the data more interpretable and facilitates operations like joining and merging.

```python
# DataFrame indexing and labels
df_indexed = pd.DataFrame(
    {'Value1': [10, 20, 30], 'Value2': [100, 200, 300]},
    index=['ID1', 'ID2', 'ID3']
)
print("\nDataFrame with custom index:")
print(df_indexed)
```

The output shows the DataFrame with a custom index. Using meaningful indices can significantly improve the readability and interpretability of the data.
A common mistake is not taking advantage of custom indices. A quick tip is to use meaningful indices to make the data more interpretable and to facilitate data manipulation.

### Set Column as Index
Setting a column as the index of a DataFrame allows for hierarchical or multi-level operations.

```python
# Set column as index
df_with_index = df.set_index('Name')
print("\nDataFrame with 'Name' as index:")
print(df_with_index)
```

The output displays the DataFrame with the specified column set as the index. This operation is useful for creating hierarchical or multi-level data structures.
A common mistake is not verifying that the column has unique values before setting it as the index. A quick tip is to ensure the column has unique values to avoid unexpected behavior.

### Reset Index
Resetting the index of a DataFrame converts the index back into a column, which can be useful after setting a column as the index.

```python
# Reset index
df_reset = df_with_index.reset_index()
print("\nAfter reset_index():")
print(df_reset)
```

The output shows the DataFrame after resetting the index. This operation is useful for converting the index back into a regular column.
A common mistake is not understanding the implications of resetting the index. A quick tip is to be aware of how resetting the index affects the structure and content of the DataFrame.

## BASIC STATISTICS
Understanding the distribution of the data is crucial before proceeding with machine learning modeling.

```python
# Basic statistics
print("\nDescriptive statistics:")
print(df.describe())  # Auto-calculates mean, std, min, max, quartiles for numeric columns
```

The output displays descriptive statistics for the numeric columns in the DataFrame. These statistics provide insights into the distribution of the data.
A common mistake is not exploring the data distribution before modeling. A quick tip is to always examine the descriptive statistics to understand the characteristics of the data.

### Statistics for Specific Columns
Calculating statistics for specific columns allows for a more detailed understanding of the data distribution.

```python
# Statistics for specific columns
print("\nMean age:", df['Age'].mean())
print("Max salary:", df['Salary'].max())
print("Standard deviation of age:", df['Age'].std())
```

The output shows the mean age, maximum salary, and standard deviation of age. These statistics provide specific insights into the distribution of the data.
A common mistake is not selecting the correct column for statistical calculations. A quick tip is to verify the column selection to ensure the correct statistics are calculated.

## DATAFRAME OPERATIONS
DataFrames support various operations for transforming and manipulating data.

```python
# Add new column
df['Bonus'] = df['Salary'] * 0.1
print("\nAfter adding Bonus column:")
print(df)
```

The output displays the DataFrame after adding a new column. This operation is useful for creating derived features or target transformations.
A common mistake is not verifying the correctness of the new column. A quick tip is to always check the new column to ensure it was created as intended.

### Apply Function to Column
Applying a function to a column allows for custom transformations on each element without explicit loops.

```python
# Apply function to column
df['AgeGroup'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Senior')
print("\nAfter applying AgeGroup:")
print(df)
```

The output shows the DataFrame after applying a function to a column. This operation is useful for custom transformations without the need for explicit loops.
A common mistake is not verifying the correctness of the transformation. A quick tip is to always check the transformed column to ensure it was created as intended.

### Drop Column
Dropping a column from a DataFrame allows for the removal of unnecessary or redundant features.

```python
# Drop column
df_dropped = df.drop('AgeGroup', axis=1)
print("\nAfter dropping AgeGroup:")
print(df_dropped.head())
```

The output displays the DataFrame after dropping a column. This operation is useful for removing unnecessary features.
A common mistake is not verifying the correctness of the dropped column. A quick tip is to always check the resulting DataFrame to ensure the correct column was dropped.

## Key Takeaways
* Series are 1D labeled arrays; DataFrames are 2D labeled tables.
* Use meaningful labels for indexing and column names.
* `describe()` provides quick statistical summaries.
* `apply()` and vectorized operations transform data without loops.
* DataFrames integrate seamlessly with ML pipelines via column/index manipulation.
