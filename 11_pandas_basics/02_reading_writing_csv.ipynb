## Pandas Reading and Writing CSV Files
This notebook covers reading and writing CSV files using pandas, a crucial skill for data science and machine learning. It explains the parameters of `read_csv` and `to_csv` functions and demonstrates their usage with practical examples.

### Reading CSV Files
CSV (Comma-Separated Values) is the most common data exchange format. The `read_csv` function loads tabular data into a DataFrame for analysis and modeling.

```python
import pandas as pd
import os

# Create sample CSV in memory and read it
csv_data = """Name,Age,Salary,Department\nAlice,25,50000,Sales\nBob,30,60000,IT\nCharlie,28,55000,HR\nDiana,22,48000,Sales"""

# Write sample to temp file and read it back
temp_file = 'sample_data.csv'
with open(temp_file, 'w') as f:
    f.write(csv_data)

# Basic read
df = pd.read_csv(temp_file)
print("Read CSV (basic):")
print(df)
```

The output shows the loaded DataFrame with default settings. Common mistakes include not specifying the correct delimiter or header row. A quick tip is to use the `nrows` parameter to test the loading process with a small sample of the data.

### Important Parameters for `read_csv`
The `read_csv` function has several important parameters that can be used to customize the loading behavior.

```python
# 1. sep/delimiter: Change delimiter if not comma
df_sep = pd.read_csv(temp_file, sep=',')

# 2. header: Specify which row contains column names
df_header = pd.read_csv(temp_file, header=0)

# 3. names: Rename columns while reading
df_names = pd.read_csv(temp_file, names=['ID', 'Name', 'Age', 'Salary'])

# 4. index_col: Set a column as index
df_index = pd.read_csv(temp_file, index_col='Name')

# 5. usecols: Load only specific columns
df_usecols = pd.read_csv(temp_file, usecols=['Name', 'Age', 'Salary'])

# 6. dtype: Specify data types for columns
df_dtype = pd.read_csv(temp_file, dtype={'Age': 'int32', 'Salary': 'float32'})

# 7. nrows: Load only first N rows
df_nrows = pd.read_csv(temp_file, nrows=2)

# 8. skiprows: Skip initial rows
df_skiprows = pd.read_csv(temp_file, skiprows=1)

# 9. na_values: Specify additional strings representing missing values
df_na_values = pd.read_csv(temp_file, na_values=['NA', 'N/A', '-'])

# 10. parse_dates: Treat columns as dates
df_parse_dates = pd.read_csv(temp_file, parse_dates=['Name'])
```

These parameters can be used to handle different file formats and requirements. For example, using the `sep` parameter can help when dealing with tab-separated or semicolon-separated files. A common mistake is not specifying the correct delimiter, which can lead to incorrect loading of the data. A quick tip is to use the `nrows` parameter to test the loading process with a small sample of the data.

### Writing CSV Files
The `to_csv` function is used to export DataFrames for use in other tools or for backup.

```python
# Basic to_csv
df.to_csv('output.csv', index=False)
print("\nDataFrame saved to 'output.csv'")
```

The output shows that the DataFrame has been saved to a CSV file. A common mistake is not specifying the correct file path or name. A quick tip is to use the `index=False` parameter to exclude the row indices from the output.

### Important Parameters for `to_csv`
The `to_csv` function has several important parameters that can be used to control the output format and content.

```python
# 1. index: Include row indices in output
df.to_csv('output_index.csv', index=True)

# 2. columns: Select which columns to save
df.to_csv('output_columns.csv', columns=['Name', 'Salary'])

# 3. sep: Use different delimiter
df.to_csv('output_sep.csv', sep=';')

# 4. na_rep: Represent missing values as specific string
df.to_csv('output_na_rep.csv', na_rep='NA')

# 5. header: Include column names in output
df.to_csv('output_header.csv', header=False)

# 6. mode: Write ('w') or append ('a') to file
df.to_csv('output_mode.csv', mode='a', header=False)
```

These parameters can be used to customize the output format and content. For example, using the `index` parameter can help when including or excluding row indices from the output. A common mistake is not specifying the correct file mode, which can lead to overwriting or appending the wrong data. A quick tip is to use the `na_rep` parameter to handle missing values consistently across systems.

### Verify and Inspect Loaded Data
Immediately check for issues after loading.

```python
# Verify and inspect loaded data
df_loaded = pd.read_csv(temp_file)
print("\n--- Data Inspection ---")
print("Shape:", df_loaded.shape)
print("Columns:", df_loaded.columns.tolist())
print("Data types:", df_loaded.dtypes.to_dict())
print("Missing values:\n", df_loaded.isnull().sum())
print("\nFirst few rows:")
print(df_loaded.head())
```

The output shows the loaded data inspection. A common mistake is not checking for missing values or data types. A quick tip is to use the `isnull` method to identify missing values and the `dtypes` attribute to verify data types.
