## Pandas Indexing - loc vs iloc
This notebook covers the basics of Pandas indexing using loc and iloc. It explains why efficient data selection is critical for feature engineering, creating subsets for train/test splits, and sample inspection during model debugging. Understanding both loc and iloc is essential for data manipulation in AI and ML workflows.

### Importing Libraries
Import the necessary libraries: pandas for data manipulation and numpy for numerical operations.

```python
import pandas as pd
import numpy as np
```

### Creating a Sample DataFrame
Create a sample DataFrame to demonstrate indexing methods.

```python
df = pd.DataFrame({
    'StudentID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Math': [85, 92, 88, 95, 78],
    'English': [90, 88, 85, 92, 95],
    'Science': [88, 85, 90, 88, 85]
})
print("Sample DataFrame:")
print(df)
```

### Output Explanation
The output displays the sample DataFrame with student information and scores.
Common Mistake: Forgetting to import necessary libraries.
Quick Tip: Use the `head()` method to view the first few rows of the DataFrame.

### ILOC: INTEGER-LOCATION BASED INDEXING
Iloc is used for position-based indexing, similar to NumPy array indexing.

```python
# Select single row by position
row_2 = df.iloc[1]  # Get second row (index 1)
print("Single row (iloc[1]):")
print(row_2)
```

### Output Explanation
The output displays the second row of the DataFrame.
Common Mistake: Confusing iloc with loc indexing.
Quick Tip: Use iloc for position-based indexing when the index is not meaningful.

### Select Multiple Rows by Position
Use slicing or lists to select multiple rows.

```python
# Select multiple rows by position
rows_0_2 = df.iloc[0:3]  # Rows at positions 0, 1, 2 (stop is exclusive)
print("Rows (iloc[0:3]):")
print(rows_0_2)
```

### Output Explanation
The output displays the first three rows of the DataFrame.
Common Mistake: Forgetting that the stop index is exclusive.
Quick Tip: Use iloc for selecting multiple rows by position.

### Select Specific Rows and Columns by Position
Use iloc to extract a rectangular subset of the DataFrame.

```python
# Select specific rows and columns by position
subset = df.iloc[1:4, 1:3]  # Rows 1-3, columns 1-2 (Name, Math)
print("Subset (iloc[1:4, 1:3]):")
print(subset)
```

### Output Explanation
The output displays a subset of the DataFrame with specific rows and columns.
Common Mistake: Confusing the order of rows and columns in iloc.
Quick Tip: Use iloc for selecting specific rows and columns by position.

### Select Single Element by Position
Use iloc to access a specific cell value.

```python
# Select single element by position
value = df.iloc[2, 3]  # Row 2, Column 3 (Charlie's English score)
print("Single element (iloc[2, 3]):", value)  # 85
```

### Output Explanation
The output displays the value of the cell at row 2, column 3.
Common Mistake: Forgetting that indexing starts at 0.
Quick Tip: Use iloc for accessing single elements by position.

### Select Last Rows/Columns
Use negative indices to access the last rows or columns.

```python
# Select last row
last_row = df.iloc[-1]  # Last row
print("Last row (iloc[-1]):")
print(last_row)
```

### Output Explanation
The output displays the last row of the DataFrame.
Common Mistake: Forgetting that negative indices start from the end.
Quick Tip: Use negative indices to access the last rows or columns.

### Fancy Indexing with Lists
Use lists to select non-contiguous rows or columns.

```python
# Fancy indexing with lists
fancy = df.iloc[[0, 2, 4], [1, 3]]  # Rows 0, 2, 4 and columns 1, 3
print("Fancy indexing (iloc[[0,2,4], [1,3]]):")
print(fancy)
```

### Output Explanation
The output displays a subset of the DataFrame with non-contiguous rows and columns.
Common Mistake: Forgetting to use lists for fancy indexing.
Quick Tip: Use lists for selecting non-contiguous rows or columns.

### LOC: LABEL-BASED INDEXING
Loc is used for label-based indexing, which is more intuitive and self-documenting than positional indexing.

```python
# Select single column by name
names = df.loc[:, 'Name']  # All rows, column 'Name'
print("'Name' column (loc[:, 'Name']):")
print(names)
```

### Output Explanation
The output displays the 'Name' column of the DataFrame.
Common Mistake: Forgetting to use the colon (:) for selecting all rows.
Quick Tip: Use loc for label-based indexing when the column name is known.

### Select Specific Row by Label
Use loc to select a row by its index label.

```python
# Select specific row by label
df_labeled = df.set_index('StudentID')
student_s003 = df_labeled.loc['S003']  # Row with index 'S003'
print("Student S003 data (loc['S003']):")
print(student_s003)
```

### Output Explanation
The output displays the row with index 'S003'.
Common Mistake: Forgetting to set the index before using loc.
Quick Tip: Use loc for selecting specific rows by label when the index is meaningful.

### Select Multiple Columns by Name
Use loc to select multiple columns by their names.

```python
# Select multiple columns by name
scores = df.loc[:, ['Math', 'English', 'Science']]
print("Scores columns:")
print(scores)
```

### Output Explanation
The output displays the 'Math', 'English', and 'Science' columns of the DataFrame.
Common Mistake: Forgetting to use a list for selecting multiple columns.
Quick Tip: Use loc for selecting multiple columns by name.

### Select Rows and Columns by Labels
Use loc to extract a subset of the DataFrame with specific rows and columns.

```python
# Select rows and columns by labels
subset_loc = df_labeled.loc[['S002', 'S004'], ['Math', 'Science']]
print("Subset (loc[['S002','S004'], ['Math','Science']]):")
print(subset_loc)
```

### Output Explanation
The output displays a subset of the DataFrame with specific rows and columns.
Common Mistake: Forgetting to use lists for selecting multiple rows or columns.
Quick Tip: Use loc for selecting rows and columns by labels.

### Slicing with Labels (Inclusive on Both Ends)
Use loc to slice the DataFrame with labels, which includes the stop label.

```python
# Slicing with labels (inclusive on both ends)
df_indexed = df.set_index('Name')
slice_loc = df_indexed.loc['Alice':'Charlie']  # Includes 'Charlie'
print("Slice 'Alice' to 'Charlie' (inclusive):")
print(slice_loc)
```

### Output Explanation
The output displays the slice of the DataFrame from 'Alice' to 'Charlie'.
Common Mistake: Forgetting that loc slicing includes the stop label.
Quick Tip: Use loc for slicing with labels when the index is meaningful.

### BOOLEAN INDEXING WITH LOC
Use loc to filter rows based on conditions.

```python
# Single condition
high_math = df.loc[df['Math'] > 85]  # Rows where Math score > 85
print("Rows where Math > 85:")
print(high_math)
```

### Output Explanation
The output displays the rows where the Math score is greater than 85.
Common Mistake: Forgetting to use the loc method for boolean indexing.
Quick Tip: Use loc for filtering rows based on conditions.

### Multiple Conditions
Use the bitwise operators (&, |, ~) to combine conditions.

```python
# Multiple conditions
good_students = df.loc[(df['Math'] > 85) & (df['English'] > 85)]  # AND
print("Rows where Math > 85 AND English > 85:")
print(good_students)
```

### Output Explanation
The output displays the rows where both Math and English scores are greater than 85.
Common Mistake: Forgetting to use parentheses for grouping conditions.
Quick Tip: Use the bitwise operators to combine conditions for filtering rows.

### OR Condition
Use the bitwise OR operator (|) to select rows where any condition is true.

```python
# OR condition
top_performers = df.loc[(df['Math'] > 90) | (df['English'] > 90)]
print("Rows where Math > 90 OR English > 90:")
print(top_performers)
```

### Output Explanation
The output displays the rows where either Math or English score is greater than 90.
Common Mistake: Forgetting to use the bitwise OR operator for combining conditions.
Quick Tip: Use the bitwise OR operator to select rows where any condition is true.

### Invert Condition with NOT
Use the bitwise NOT operator (~) to exclude rows that meet a condition.

```python
# Invert condition with NOT
below_90_math = df.loc[~(df['Math'] > 90)]  # NOT Math > 90
print("Rows where NOT (Math > 90):")
print(below_90_math)
```

### Output Explanation
The output displays the rows where the Math score is not greater than 90.
Common Mistake: Forgetting to use the bitwise NOT operator for inverting conditions.
Quick Tip: Use the bitwise NOT operator to exclude rows that meet a condition.

### Using isin for Membership
Use the isin method to filter rows based on a list of allowed values.

```python
# Using isin for membership
names_to_select = ['Alice', 'Bob', 'Eve']
selected = df.loc[df['Name'].isin(names_to_select)]
print("Rows where Name in ['Alice', 'Bob', 'Eve']:")
print(selected)
```

### Output Explanation
The output displays the rows where the Name is in the list ['Alice', 'Bob', 'Eve'].
Common Mistake: Forgetting to use the isin method for membership testing.
Quick Tip: Use the isin method to filter rows based on a list of allowed values.

### LOC WITH AT AND IAT (SINGLE ELEMENT)
Use at and iat for efficient single-element access.

```python
# at: single element by label
charlie_math = df_labeled.at['S003', 'Math']
print(f"Charlie's Math score (at): {charlie_math}")
```

### Output Explanation
The output displays Charlie's Math score.
Common Mistake: Forgetting to use at for single-element access by label.
Quick Tip: Use at for efficient single-element access by label.

### IAT: SINGLE ELEMENT BY POSITION
Use iat for efficient single-element access by position.

```python
# iat: single element by position
elem = df.iat[2, 3]  # Row 2, Column 3
print(f"Element at position [2, 3] (iat): {elem}")
```

### Output Explanation
The output displays the element at position [2, 3].
Common Mistake: Forgetting to use iat for single-element access by position.
Quick Tip: Use iat for efficient single-element access by position.

### PRACTICAL MACHINE LEARNING SCENARIOS
Apply indexing techniques to real-world machine learning scenarios.

```python
# Scenario 1: Create train-test split
train_idx = [0, 1, 2]
test_idx = [3, 4]
X_train = df.iloc[train_idx, 1:5]  # Columns 1-4 (features)
X_test = df.iloc[test_idx, 1:5]
print("Train set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

### Output Explanation
The output displays the shape of the train and test sets.
Common Mistake: Forgetting to use iloc for selecting rows and columns by position.
Quick Tip: Use iloc for creating train-test splits by selecting rows and columns by position.

## Key Takeaways
* Iloc: Use for position-based indexing (like NumPy).
* Loc: Use for label-based indexing (more intuitive and readable).
* Loc supports boolean indexing for filtering; iloc does not.
* Use at/iat for single-element access for efficiency.
* Mastering both enables effective data manipulation for ML workflows.
