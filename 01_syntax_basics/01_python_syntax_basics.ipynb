### Printing Output
The `print()` function is used to output text or values to the screen. It is a basic function in Python and is often used for debugging purposes. The syntax pattern is `print('output')` or `print(variable)`.

````python
print("Hello world")
````

The output of this cell is simply the string 'Hello world'. A common mistake is to forget the quotes around the string. A quick tip is to use the `print()` function to test the output of your code.

### Variables and Assignment
In Python, variables are used to store values. The assignment operator `=` is used to assign a value to a variable. The syntax pattern is `variable = value`.

````python
name = "sourav"
age = 19
value = 10.99
wife = None
````

The variables `name`, `age`, `value`, and `wife` are assigned the values 'sourav', 19, 10.99, and None, respectively. A common mistake is to use a reserved keyword as a variable name. A quick tip is to use descriptive variable names to improve code readability.

````python
print("My name is: ", name)
print("My age is:", age)
print("My value is:", value)
````

The output of this cell is the values of the variables `name`, `age`, and `value`. A common mistake is to forget to include the variable name in the `print()` function. A quick tip is to use the `print()` function to test the values of variables.

The output of this cell is the data type of each variable. A common mistake is to assume that the data type of a variable is different from what it actually is. A quick tip is to use the `type()` function to determine the data type of a variable.

## Data Structures
This section covers the fundamental data structures in Python: Lists, Tuples, Sets, and Dictionaries. Understanding these is crucial for handling and manipulating data effectively in any AI or Data Science project.

### 💡 Lists
**Concept** — Lists are Python's dynamic arrays. They are ordered, mutable (changeable), and can hold different data types. They are ideal for storing sequences of items that might need to change during program execution.

**AI/ML Connection** — Lists are workhorses in AI for collecting results during model training, like storing the loss value after each epoch, or for holding batches of tokenized sentences before they are converted to numerical tensors.

````python
# Create a list of features
features = [0.5, 0.8, 0.1]

# Add an element to the end (O(1) operation)
features.append(0.9)
print(f"After append: {features}")

# Insert an element at a specific position (O(n) operation)
features.insert(0, 1.0)
print(f"After insert: {features}")

# Remove and get the last element
popped_value = features.pop()
print(f"Popped value: {popped_value}, List is now: {features}")

# Sort the list in place
features.sort(reverse=True)
print(f"Sorted list: {features}")
````

**⚠️ Common Mistake**: Modifying a list while iterating over it. This can cause the loop to skip elements unexpectedly because the list's indices shift. Always iterate over a *copy* if you need to remove items inside the loop (e.g., `for item in my_list[:]`).

**✅ Key Takeaway**: Use lists when you need an ordered collection of items that may grow, shrink, or change.

### 💡 Tuples
**Concept** — A tuple is an ordered, **immutable** (unchangeable) collection of elements. Once created, you cannot add, remove, or change its elements. This makes them slightly more memory-efficient and faster than lists for fixed data.

**AI/ML Connection** — Tuples are perfect for defining fixed structures, like the shape of a tensor `(3, 224, 224)`, RGB color values, or a set of hyperparameters that should not change during training. Their immutability also allows them to be used as keys in a dictionary.

````python
# A tuple representing the shape of an image tensor (Channels, Height, Width)
tensor_shape = (3, 224, 224)

# You cannot change an element in a tuple
# tensor_shape[0] = 1  # This would raise a TypeError

# Accessing elements is the same as with lists
num_channels = tensor_shape[0]
total_pixels = tensor_shape[1] * tensor_shape[2]
print(f"Number of channels: {num_channels}, Total pixels: {total_pixels}")

# A single-element tuple MUST have a trailing comma
single_element_tuple = (5,)
not_a_tuple = (5)
print(f"This is a tuple: {type(single_element_tuple)}")
print(f"This is just an integer: {type(not_a_tuple)}")
````

**⚠️ Common Mistake**: Forgetting the trailing comma when creating a tuple with only one element. `(5)` is just the integer 5, but `(5,)` is a tuple containing the integer 5.

**✅ Key Takeaway**: Use tuples for fixed collections of data that should not be modified. Their immutability makes them reliable for use as dictionary keys.

### 💡 Sets
**Concept** — Sets are **unordered**, mutable collections that only store **unique, hashable** elements. They are highly optimized for checking if an element is present in the collection (membership testing) due to their underlying hash table structure.

**AI/ML Connection** — Sets are essential in Natural Language Processing (NLP) for creating a vocabulary of all unique words from a large text dataset. They automatically handle duplicate words for you. They are also used to find unique labels in a classification problem.

````python
# A set of predicted classes, duplicates are automatically removed
predicted_classes = {0, 1, 1, 2, 0, 3, 1}
print(f"Unique predicted classes: {predicted_classes}")

# Add a new element
predicted_classes.add(4)
print(f"After adding 4: {predicted_classes}")

# Remove an element safely using discard() (does not raise an error if missing)
predicted_classes.discard(0)
print(f"After discarding 0: {predicted_classes}")

# Remove an element using remove() (raises a KeyError if the element doesn't exist)
# predicted_classes.remove(99) # This would cause a KeyError

# Check for membership (this is a very fast O(1) operation in a set)
print(f"Is class 2 in predictions? {2 in predicted_classes}")
````

**⚠️ Common Mistake**: Creating an empty set with `{}`. This syntax actually creates an empty **dictionary**. To create an empty set, you must use the `set()` constructor.

**✅ Key Takeaway**: Use sets when you need to store a collection of unique items and the order doesn't matter. They are your go-to for fast membership testing and removing duplicates.

### 💡 Dictionaries
**Concept** — Dictionaries store data in **key-value pairs**. Keys must be unique and immutable (like strings, numbers, or tuples). Values can be anything. As of Python 3.7, dictionaries maintain the order in which items were inserted.

**AI/ML Connection** — Dictionaries are the backbone of configuration management. They are used to define hyperparameter grids for model tuning (e.g., `{'learning_rate': 0.01, 'epochs': 100}`), to map feature names to their values, and to structure JSON data returned by APIs.

````python
# A dictionary for model hyperparameters
hyperparameters = {"learning_rate": 0.01, "batch_size": 32}

# Add a new key-value pair
hyperparameters["epochs"] = 100
print(f"Updated hyperparameters: {hyperparameters}")

# Access a value by its key
lr = hyperparameters["learning_rate"]
print(f"Learning Rate: {lr}")

# Safely get a value using .get(), which provides a default if the key is not found
momentum = hyperparameters.get("momentum", 0.9) # "momentum" doesn't exist, so it returns 0.9
print(f"Momentum (default value): {momentum}")

# Get all keys and values
print(f"Keys: {hyperparameters.keys()}")
print(f"Values: {hyperparameters.values()}")
print(f"Items: {hyperparameters.items()}")
````

**⚠️ Common Mistake**: Using a mutable object like a list as a dictionary key. This will raise a `TypeError` because keys must be hashable (their value can't change).

**✅ Key Takeaway**: Use dictionaries when you need to store and retrieve data based on a specific name or identifier (the key), rather than by its position. They are perfect for creating flexible, readable data structures.

### 💡 Comprehensions (List / Dict / Set)
**Concept** — Comprehensions are a concise, C-optimized syntax to generate collections from iterables. They are often more readable and performant than an equivalent `for` loop, allowing you to create a new collection in a single line of code.

**AI/ML Connection** — Comprehensions are heavily used in data preprocessing pipelines for tasks like applying a transformation to every element in a feature vector, filtering out low-quality data points, or building lookup dictionaries.

````python
raw_data = [1, 2, -3, 4, -5, 2]

# List comprehension: Square only the positive numbers
squares = [x**2 for x in raw_data if x > 0]
print(f"Squares of positive numbers: {squares}")

# Set comprehension: Get the unique absolute values
abs_set = {abs(x) for x in raw_data}
print(f"Unique absolute values: {abs_set}")

# Dictionary comprehension: Map each negative number to its absolute value
neg_map = {x: abs(x) for x in raw_data if x < 0}
print(f"Negative value map: {neg_map}")
````

**⚠️ Common Mistake**: Writing overly complex, nested comprehensions that are hard to read. If a comprehension needs multiple lines or complex logic (e.g., nested `if/else`), a standard `for` loop is often more maintainable and easier to debug.

**✅ Key Takeaway**: Use comprehensions to create new collections from existing ones in a single, readable line of code. They are a hallmark of idiomatic Python.

### 💥 Gotchas — Data Structures

```python
# GOTCHA 1: List multiplication creates shared references, not independent copies.
# This is a very common bug for beginners creating 2D grids.
matrix = [[0] * 3] * 3      # All 3 rows point to the SAME list object in memory
matrix[0][0] = 1            # This corrupts ALL rows: [[1,0,0],[1,0,0],[1,0,0]]

# The CORRECT way is to use a comprehension to create independent inner lists.
correct_matrix = [[0 for _ in range(3)] for _ in range(3)]
correct_matrix[0][0] = 1    # Only affects the first row: [[1,0,0],[0,0,0],[0,0,0]]

# GOTCHA 2: Tuple hashability depends on ALL its elements being immutable.
# A tuple can only be a dictionary key if it contains no mutable objects.
bad_key = (1, [2, 3])       # This will raise a TypeError because the list inside is mutable.
# my_dict = {bad_key: "value"} # Uncommenting this line would cause an error.

# GOTCHA 3: Dictionary key collisions — the last one wins.
# When creating a dictionary, if you use the same key multiple times, the last value assigned will overwrite previous ones.
merged = {'a': 1, 'b': 2, 'a': 3} # The final value for 'a' will be 3.
print(merged) # Output: {'a': 3, 'b': 2}
```

### 📋 Cheat Sheet — Data Structures

```python
# List: Ordered, mutable, allows duplicates. For dynamic data.
L = [1, "a", 1]
L.append(2)
L.pop()
L.sort()

# Tuple: Ordered, IMMUTABLE, allows duplicates. For fixed data.
T = (1, "a", 1)
# T[0] = 2 # This would be a TypeError

# Set: Unordered, mutable, NO duplicates. For uniqueness and math operations.
S = {1, "a"}
S.add(2)
S.discard(1)

# Dictionary: Key-Value pairs, keys are unique & immutable. For lookups by name.
D = {'key1': 'value1', 'key2': 'value2'}
D['key3'] = 'value3'
print(D.get('key1'))

# Comprehensions: Concise way to create collections.
squares = [x**2 for x in range(5)] # List comprehension
unique_squares = {x**2 for x in range(5)} # Set comprehension
square_map = {x: x**2 for x in range(5)} # Dictionary comprehension
```
