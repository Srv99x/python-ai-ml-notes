## Escape Sequences in Python Strings

This notebook covers the use of escape sequences in Python strings. Escape sequences are special commands inside a string that allow you to insert characters that are otherwise difficult or impossible to type directly, such as newlines, tabs, or quotes within the string itself.

### What are Escape Sequences?

An escape sequence is a backslash `\` followed by a character. The backslash tells the Python interpreter that the next character has a special meaning.

For example, if you want to include a double quote inside a string that is enclosed in double quotes, you must "escape" it.

```python
# Without escaping, this would be a syntax error
quote = "He said, \"Python is amazing!\""
print(quote)
```

### Common Escape Sequences

Here are some of the most frequently used escape sequences:

```python
print("Hello\\nWorld")      # \n: Newline
print("Column 1\\tColumn 2") # \t: Tab
print("This is a backslash: \\\\") # \\: Backslash
print('It\\'s a beautiful day.') # \': Single Quote
```

### 💥 Gotcha: Raw Strings

Sometimes, you don't want Python to interpret escape sequences. This is common when working with file paths on Windows, which use backslashes.

To prevent this, you can use a **raw string** by adding an `r` before the opening quote. In a raw string, backslashes are treated as literal characters.

```python
# Without a raw string, \U in \Users can be misinterpreted as a Unicode escape
path = "C:\\Users\\Sourav\\Desktop"
print(f"Problematic Path: {path}")

# With a raw string, the path is correct
raw_path = r"C:\Users\Sourav\Desktop"
print(f"Correct Path: {raw_path}")
```

### 📋 Cheat Sheet — Common Escape Sequences

| Sequence | Meaning |
| :--- | :--- |
| `\\n` | Newline |
| `\\t` | Horizontal Tab |
| `\\\\` | Literal Backslash |
| `\\'` | Single Quote |
| `\\\"` | Double Quote |
| `\\r` | Carriage Return |
| `\\b` | Backspace |
