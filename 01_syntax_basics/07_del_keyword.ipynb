## The `del` Keyword and Memory Management

This notebook covers the `del` keyword inside Python. While it seems simple—it "deletes" things—understanding exactly how it works is crucial for memory management, especially in data-intensive fields like AI and Machine Learning.

### What Does `del` Actually Do?

**Concept** — In Python, `del` is a statement that removes a **reference** to an object, not necessarily the object itself. 

Think of a variable name as a sticky note attached to a balloon (the object in memory). `del variable_name` rips the sticky note off. 

If a balloon has no sticky notes left attached to it, Python's **Garbage Collector** will eventually come along and pop the balloon, freeing up the memory.

**AI/ML Connection** — When training large deep learning models or processing massive Pandas DataFrames, your computer's RAM or your GPU's VRAM can fill up quickly. Using `del` to remove references to variable names holding giant datasets after you are done with them allows Python to free up that memory, preventing Out-Of-Memory (OOM) crashes.

```python
# Example 1: Deleting a simple variable reference
large_data = [x for x in range(1_000_000)]  # Imagine this takes a lot of memory

print(f"Data length: {len(large_data)}")

# Remove the reference
del large_data

# The next line will cause a NameError because the sticky note is gone
try:
    print(large_data)
except NameError as e:
    print(f"Error: {e}")
```

### Deleting Items from Collections

You can also use `del` to remove specific items from mutable collections like Lists or Dictionaries.

```python
# Deleting an item from a list by index
fruits = ["Apple", "Banana", "Cherry", "Date"]
del fruits[1]  # Removes Banana
print(f"List after del: {fruits}")

# Deleting a slice from a list
del fruits[1:3] # Removes index 1 and 2
print(f"List after slice del: {fruits}")

# Deleting from a dictionary by key
hyperparams = {"lr": 0.01, "batch_size": 32, "dropout": 0.5}
del hyperparams["dropout"]
print(f"Dict after del: {hyperparams}")
```

### Deleting Object Attributes

In Object-Oriented Programming, `del` can remove an attribute from an instance of a class.

```python
class ModelConfig:
    def __init__(self):
        self.optimizer = "Adam"
        self.learning_rate = 0.001

config = ModelConfig()
print(f"Before: {config.__dict__}")

# Delete the learning_rate attribute from this specific instance
del config.learning_rate

print(f"After: {config.__dict__}")

# Trying to access it now will raise an AttributeError
try:
    print(config.learning_rate)
except AttributeError as e:
    print(f"Error: {e}")
```

### 💥 Gotcha: `del` vs `.remove()` vs `.pop()` for Lists

It's easy to confuse `del` with list methods.
*   **`del list[index]`**: A Python *statement*, deletes by **index**. Does not return the deleted value.
*   **`list.pop(index)`**: A list *method*, deletes by **index**. **Returns** the deleted value so you can use it.
*   **`list.remove(value)`**: A list *method*, deletes the first occurrence of a specific **value**, regardless of its index.

**✅ Key Takeaway**: Use `del` when you need to free up memory from large variables, remove slices from lists, or clean up object attributes. But for normal list manipulation, `.pop()` or `.remove()` are often more useful.
