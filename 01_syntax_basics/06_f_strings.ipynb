## Formatted String Literals (f-strings)

This notebook covers f-strings, the modern, powerful, and preferred way to format strings in Python. They allow you to embed expressions directly inside string literals, making your code more readable and concise.

### What are f-strings?

An f-string is a string literal that is prefixed with the letter `f` or `F`. Inside the string, you can place any valid Python expression inside curly braces `{}`. Python will evaluate the expression and embed the result directly into the string.

This is much cleaner than older methods like string concatenation (`+`) or the `.format()` method.

```python
name = "Alice"
age = 30
pi = 3.14159

# Basic f-string usage
print(f"My name is {name} and I am {age} years old.")

# You can put any valid expression inside the braces
print(f"In 5 years, I will be {age + 5} years old.")

# You can even call functions
print(f"My name in uppercase is {name.upper()}.")

# You can format numbers
print(f"The value of pi to 2 decimal places is {pi:.2f}.")
```

### 💥 Gotcha: The "Debugging" f-string (Python 3.8+)

A fantastic feature introduced in Python 3.8 is the ability to quickly print a variable's name and its value for debugging. Simply add an equals sign `=` after the expression.

```python
user_status = "active"
login_count = 42

# Typical way to print variables for debugging
print(f"user_status='{user_status}', login_count={login_count}")

# The magical Python 3.8+ debugging syntax: just add an '='
print(f"{user_status=} {login_count=}")
```

## 📋 Cheat Sheet — f-strings

| Feature | Syntax Example | Notes |
| :--- | :--- | :--- |
| **Basic Interpolation** | `f"Hello {name}"` | Embeds variables directly. |
| **Math Operations**| `f"Total: {price * qty}"` | Evaluates inline math. |
| **Function Calls** | `f"Caps: {text.upper()}"` | Runs methods inline. |
| **Decimal Formatting** | `f"Score: {0.876:.2f}"` | Rounds to 2 decimal places. |
| **Whitespace Padding** | `f">{text:>10}<"` | Right-aligns within 10 spaces. |
| **Debugging Print**| `f"{variable=}"` | Prints variable name and its value. |
