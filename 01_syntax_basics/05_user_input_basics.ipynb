## Taking User Input and Type Conversion

This notebook covers the essential skill of reading input from a user and converting it into the correct data type for processing. This is a fundamental building block for creating interactive programs.

### The Golden Rule of `input()`: It Always Returns a String!

The most important thing to remember about the `input()` function is that it **always** gives you back a string, regardless of what the user types.

If you ask for a number and the user types `123`, `input()` does not give you the number 123, it gives you the string `"123"`.

```python
# The input() function pauses the program and waits for the user to type something
age_str = input("Please enter your age: ")

print(f"The data type of the input is: {type(age_str)}")

# If we try to do math with this string, we get an error
try:
    age_str + 10
except TypeError as e:
    print(f"Error when trying to add 10 to the string: {e}")
```

### Type Casting: Converting The Input

To perform mathematical operations, you must convert the string from `input()` to a numeric type. This is called **type casting**.

- `int()`: Converts a string to an integer.
- `float()`: Converts a string to a floating-point number (a number with a decimal).

```python
# Get the input as a string
age_str = input("Enter your age again: ")

# Cast the string to an integer
age_int = int(age_str)
print(f"The data type is now: {type(age_int)}")
print(f"In ten years, you will be: {age_int + 10}")

# You can also do it in one line
price_float = float(input("Enter the price of the item: "))
print(f"The price with tax is: {price_float * 1.2}")
```

### 💥 Gotcha: Handling Invalid Input

What happens if the user types "hello" when you ask for their age?

The `int()` function can't convert "hello" to a number, so it will raise a `ValueError`, and your program will crash.

A robust program must anticipate this! The correct way to handle user input that needs to be a number is to use a `try...except` block.

```python
# This is the standard, safe way to get a number from a user

while True: # Loop forever until we get valid input
    age_str = input("Please enter your age (and be honest!): ")
    try:
        # Try to convert the input to an integer
        age_int = int(age_str)
        # If the conversion succeeds, break out of the loop
        break
    except ValueError:
        # If int() raises a ValueError, the input was not a valid number
        print("Invalid input. Please enter a whole number.")

print(f"Success! You entered the number: {age_int}")
```

## 📋 Cheat Sheet — User Input

| Step | Action | Code Example |
| :--- | :--- | :--- |
| **1. Prompt User** | Use `input()` with a clear prompt. | `name = input("Enter your name: ")` |
| **2. Receive String** | Remember `input()` always returns a string. | `num_str = input("Enter a number: ")` |
| **3. Validate & Cast** | Use `try-except` to safely convert to the target type. | `try: num = int(num_str) except ValueError: print("Invalid")` |
| **4. Use Value** | Once converted, you can use the variable in your logic. | `print(num * 2)` |
