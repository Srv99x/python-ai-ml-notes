## Advanced String Methods and Techniques

This notebook explores more advanced and specialized string methods. These are crucial for data validation, cleaning, and complex text manipulation, which are common tasks in data science and AI.

### 💡 Searching, Counting, and Replacing

These methods help you locate, quantify, and change substrings within a larger string.

**AI/ML Connection**: These are vital for feature engineering. You might `.count()` the number of times a certain keyword appears in a product review as a feature for a sentiment analysis model, or `.replace()` placeholder tokens in a text template.

```python
log_entry = "INFO:2024-03-24:user_login_success:user_id=123"

# .startswith() - Check for a prefix
print(f"Is this a log entry? {log_entry.startswith('INFO')}")

# .endswith() - Check for a suffix
print(f"Does the log mention user 123? {log_entry.endswith('user_id=123')}")

# .find() - Find the starting index of a substring
# Returns -1 if the substring is not found
print(f"Index of 'user_login': {log_entry.find('user_login')}")
print(f"Index of 'error': {log_entry.find('error')}")

# .count() - Count non-overlapping occurrences of a substring
print(f"Count of 'user': {log_entry.count('user')}")

# .replace() - Replace a substring with another
# Remember: this returns a NEW string
new_log_entry = log_entry.replace('success', 'failure')
print(f"Modified log: {new_log_entry}")
```

### 💡 Validating String Content (`is...` methods)

This family of methods checks the characters within a string and returns `True` or `False`. They are extremely useful for validating user input or cleaning data before processing.

**AI/ML Connection**: Before converting a column of data to a numeric type (e.g., for a regression model), you must first ensure it doesn't contain non-numeric characters. The `.isnumeric()` method is perfect for this kind of data validation step in a pipeline.

```python
s1 = "12345"
s2 = "abcde"
s3 = "StringWithUpperAndLower"
s4 = "  "
s5 = "Has Spaces"

# .isnumeric() - True if all characters are numeric
print(f"'{s1}'.isnumeric(): {s1.isnumeric()}")

# .isalpha() - True if all characters are alphabetic
print(f"'{s2}'.isalpha(): {s2.isalpha()}")

# .isalnum() - True if all characters are alphanumeric (letters or numbers)
print(f"'{s1}'.isalnum(): {s1.isalnum()}")
print(f"'{s2}'.isalnum(): {s2.isalnum()}")
print(f"'{s5}'.isalnum(): {s5.isalnum()}") # False because of the space

# .islower() / .isupper() - Check for case
print(f"'{s3}'.islower(): {s3.islower()}")
print(f"'word'.islower(): {'word'.islower()}")

# .isspace() - True if all characters are whitespace
print(f"'{s4}'.isspace(): {s4.isspace()}")
```

### 💡 Method Chaining

Since string methods return new strings, you can "chain" them together in a single, elegant line. The output of one method becomes the input for the next, reading from left to right.

**AI/ML Connection**: Method chaining is the foundation of many data cleaning pipelines in Pandas. You'll see long chains of operations applied to entire columns of text data to normalize it efficiently.

```python
# A messy string from a hypothetical data source
raw_data = "  $$$ ITEM_ID: item_456 $$$  "

# Clean it up with method chaining
# 1. .strip() removes leading/trailing whitespace
# 2. .lower() converts to lowercase
# 3. .replace() removes the '$' signs
# 4. .removeprefix() is a newer method (Python 3.9+) to remove a specific prefix
cleaned_data = raw_data.strip().lower().replace('$', '').removeprefix(' item_id: ')

print(f"Raw: '{raw_data}'")
print(f"Cleaned: '{cleaned_data}'")
```

### 💥 Gotcha: `.find()` vs `.index()`

Both methods are used to find the starting index of a substring. The key difference is how they behave when the substring is **not found**.

*   `.find()`: Returns `-1`. This is safe and allows you to check the result with a simple `if result == -1:`.
*   `.index()`: Raises a `ValueError`. This can crash your program if you don't wrap it in a `try...except` block.

**Rule of thumb**: Use `.find()` for safe searching. Use `.index()` only when you are certain the substring exists and you want the program to fail loudly if it doesn't.

```python
text = "hello world"

# Using .find() is safe
print(f"Finding 'world' with .find(): {text.find('world')}")
print(f"Finding 'galaxy' with .find(): {text.find('galaxy')}") # Returns -1

# Using .index() can cause an error
print(f"Finding 'world' with .index(): {text.index('world')}")
try:
    print(text.index('galaxy'))
except ValueError as e:
    print(f"Finding 'galaxy' with .index() raised an error: {e}")
```

## 📋 Cheat Sheet — Advanced String Methods

| Method | Description | Example | Result |
| :--- | :--- | :--- | :--- |
| **`.startswith(sub)`** | Returns `True` if the string starts with `sub`. | `'abc'.startswith('a')` | `True` |
| **`.endswith(sub)`** | Returns `True` if the string ends with `sub`. | `'abc'.endswith('c')` | `True` |
| **`.find(sub)`** | Returns the first index of `sub`. **Returns -1 if not found.** | `'abc'.find('b')` | `1` |
| **`.index(sub)`** | Returns the first index of `sub`. **Raises `ValueError` if not found.** | `'abc'.index('b')` | `1` |
| **`.count(sub)`** | Counts non-overlapping occurrences of `sub`. | `'ababa'.count('ab')` | `2` |
| **`.capitalize()`** | Capitalizes the first character, rest lowercase. | `'hELLo'.capitalize()` | `'Hello'` |
| **`.title()`** | Capitalizes the first character of each word. | `'hello world'.title()` | `'Hello World'` |
| **`.isnumeric()`** | `True` if all characters are numeric (0-9). | `'123'.isnumeric()` | `True` |
| **`.isalpha()`** | `True` if all characters are alphabetic (a-z). | `'abc'.isalpha()` | `True` |
| **`.isalnum()`** | `True` if all characters are alphanumeric (a-z, 0-9). | `'abc123'.isalnum()` | `True` |
| **`.isspace()`** | `True` if all characters are whitespace. | `'   '.isspace()` | `True` |
| **`.islower()`** | `True` if all cased characters are lowercase. | `'abc'.islower()` | `True` |
| **`.isupper()`** | `True` if all cased characters are uppercase. | `'ABC'.isupper()` | `True` |
