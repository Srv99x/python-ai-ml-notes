## String Operations and Common Methods
This notebook covers the basics of string operations in Python, including indexing, slicing, and concatenation.
It provides a quick recap of string methods and their usage.

### Indexing
Indexing is used to access a specific character in a string. In Python, indexing starts at 0.
The syntax for indexing is `str[index]`.

```python
str = "Hello world!"
ch = str[0]  # Access the first character
print(ch)
```

The output of the above code will be `H`, which is the first character of the string.
A common mistake is to forget that indexing starts at 0, not 1.
A quick tip is to use the `len()` function to get the length of the string and avoid index out of range errors.

### Slicing
Slicing is used to extract a subset of characters from a string. The syntax for slicing is `str[start:stop]`.
If the start index is omitted, it defaults to 0. If the stop index is omitted, it defaults to the end of the string.

```python
print(str[1:4])  # Extract characters from index 1 to 3
print(str[:12])  # Extract characters from the start to index 11
print(str[0:len(str)])  # Extract all characters
print(str[0:])  # Extract all characters
```

The output of the above code will be `ell`, `Hello worl`, `Hello world!`, and `Hello world!` respectively.
A common mistake is to forget that the stop index is exclusive.
A quick tip is to use the `len()` function to get the length of the string and avoid index out of range errors.

### Negative Indexing
Negative indexing is used to access characters from the end of the string. The syntax for negative indexing is `str[-start]`.
The last character has an index of -1, the second last character has an index of -2, and so on.

```python
print(str[-12:-1])  # Extract characters from the 12th last character to the second last character
```

The output of the above code will be `Hello worl`.
A common mistake is to forget that the stop index is exclusive.
A quick tip is to use the `len()` function to get the length of the string and avoid index out of range errors.

### Concatenation
Concatenation is used to join two or more strings together. The syntax for concatenation is `str1 + str2`.
You can also use the `+` operator to concatenate strings with other data types, but you need to convert them to strings first.

```python
myStr1 = "Hello"
myStr2 = "World"
print(myStr1 + " " + myStr2)  # Concatenate two strings with a space in between
```

The output of the above code will be `Hello World`.
A common mistake is to forget to convert other data types to strings before concatenating.
A quick tip is to use the `str()` function to convert other data types to strings.

## Key Takeaways
* Indexing is used to access a specific character in a string.
* Slicing is used to extract a subset of characters from a string.
* Negative indexing is used to access characters from the end of the string.
* Concatenation is used to join two or more strings together.

### 💡 Common String Methods
Strings come with many built-in methods to perform common operations. These are essential for cleaning and preparing text data.

**AI/ML Connection**: In Natural Language Processing (NLP), you spend most of your time cleaning text. Methods like `.lower()`, `.strip()`, and `.replace()` are used constantly to normalize text before feeding it into a model.

```python
text = "  Some Text with Mixed Case and Spaces.  "

# .lower() - convert to lowercase
lower_text = text.lower()
print(f"lower(): '{lower_text}'")

# .upper() - convert to uppercase
upper_text = text.upper()
print(f"upper(): '{upper_text}'")

# .strip() - remove leading/trailing whitespace
stripped_text = text.strip()
print(f"strip(): '{stripped_text}'")

# .replace(old, new) - replace a substring
replaced_text = stripped_text.replace("Text", "Data")
print(f"replace(): '{replaced_text}'")

# .split(delimiter) - split the string into a list of substrings
words = stripped_text.split(' ')
print(f"split(): {words}")

# .startswith(prefix) - check if the string starts with a prefix
print(f"Starts with 'Some'? {stripped_text.startswith('Some')}")

# .endswith(suffix) - check if the string ends with a suffix
print(f"Ends with '.'? {stripped_text.endswith('.')}")
```

### 💥 Gotcha: Strings are Immutable!

A critical concept to remember is that strings are **immutable**. This means they **cannot be changed in place**. All string methods that seem to modify a string actually return a **new, modified copy**.

Forgetting this is a very common source of bugs.

```python
my_string = "Hello"

# This does NOT change my_string
my_string.lower()

print(f"Original string is unchanged: {my_string}")

# To change it, you must reassign the result
my_string = my_string.lower()
print(f"String after reassignment: {my_string}")

# This will cause a TypeError because you cannot change a character in place
# my_string[0] = "H"
```
