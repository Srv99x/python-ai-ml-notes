## File Handling BasicsThis topic covers the fundamentals of file handling in Python, including opening, reading, and closing files.It provides a basic understanding of how to work with files in Python.

### Opening a FileThe `open()` function is used to open a file. It returns a file object, which can be used to read or write to the file.The syntax pattern is: `file_object = open('file_name', 'mode')`

```python
# Open the file 'text.txt' in read mode ('r' is the default mode)f = open('text.txt')  # 'f' is the file object
```

The output of this cell will be the file object itself. This is not very informative, but it's a good sign that the file was opened successfully.A common mistake is trying to open a file that does not exist. Make sure the file is in the same directory as your Python script.A quick tip is to use the `with` statement to open files, which automatically closes the file when you're done with it.

### Reading a FileThe `read()` method is used to read the contents of a file. It returns a string containing the entire file.The syntax pattern is: `data = file_object.read()`

```python
# Read the contents of the filedata = f.read()  # 'data' is a string containing the file contentsprint(data)  # Print the file contents
```

The output of this cell will be the contents of the file. This could be text, numbers, or any other type of data.A common mistake is trying to read a file that is not open. Make sure to open the file before trying to read it.A quick tip is to use the `readlines()` method to read a file line by line, which can be more memory-efficient for large files.

### Closing a FileThe `close()` method is used to close a file. It's good practice to close files when you're done with them to free up system resources.The syntax pattern is: `file_object.close()`

```python
# Close the filef.close()  # Close the file object
```

The output of this cell will be None, indicating that the file was closed successfully.A common mistake is forgetting to close files. Use the `with` statement to avoid this mistake.A quick tip is to use the `try`-`finally` block to ensure that files are closed even if an error occurs.

## Key Takeaways* The `open()` function is used to open a file.* The `read()` method is used to read the contents of a file.* The `close()` method is used to close a file.* Use the `with` statement to open files, which automatically closes the file when you're done with it.
