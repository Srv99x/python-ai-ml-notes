## Writing and Appending Data to Files SafelyThis topic covers the basics of writing data to files in Python.It includes opening a file, writing to it, and safely closing it.

### Writing to a FileTo write to a file, we first need to open it in write mode ('w').The `write()` function is then used to add content to the file.

```python
# Define the string to be written to the filemy_string = "Sourav is a good boy!"# Open the file in write mode ('w')f = open("myfile.txt", "w")# Write the string to the filef.write(my_string)# Close the file to free up system resourcesf.close()
```

### Understanding the OutputAfter running this code, a new file named `myfile.txt` will be created in the current working directory.One common mistake is forgetting to close the file, which can lead to data corruption or other issues.A quick tip is to use the `with` statement for file operations, which automatically closes the file when done.

## Key Takeaways* Opening a file in write mode ('w') will overwrite any existing content.* The `write()` function is used to add content to a file.* It's essential to close a file after use to free up system resources.
